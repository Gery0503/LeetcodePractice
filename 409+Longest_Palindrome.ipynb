# 409. Longest Palindrome

[Link to Problem](https://leetcode.com/problems/longest-palindrome/)

### Description

Given a string `s` which consists of lowercase or uppercase letters, return the length of the **longest palindrome** that can be built with those letters.

Letters are **case sensitive**, for example, `"Aa"` is not considered a palindrome here.

**Example 1:**
```
Input: s = "abccccdd"
Output: 7
Explanation: One longest palindrome that can be built is "dccaccd", whose length is 7.
```

**Example 2:**
```
Input: s = "a"
Output: 1
Explanation: The longest palindrome that can be built is "a", whose length is 1.
```

**Constraints:**
- `1 <= s.length <= 2000`
- `s` consists of lowercase **and/or** uppercase English letters only.

````python
# first attempt version
from collections import Counter
def longestPalindrome(s: str) -> int:
    table = Counter(s)
    odd_taken = False
    result = 0
    for letter, count in table.items():
        if count % 2 == 1:
            if odd_taken:
                count -= 1
            else:
                odd_taken = True
        result += count
    return result

if __name__ == '__main__':
    # Test Case 1
    s1 = "abccccdd"
    result1 = longestPalindrome(s1)
    print(f"Test 1 Input: s = \"{s1}\"")
    print(f"Test 1 Output: {result1}") # Expected: 7

    # Test Case 2
    s2 = "a"
    result2 = longestPalindrome(s2)
    print(f"\nTest 2 Input: s = \"{s2}\"")
    print(f"Test 2 Output: {result2}") # Expected: 1

    # Test Case 3 (All same characters)
    s3 = "bb"
    result3 = longestPalindrome(s3)
    print(f"\nTest 3 Input: s = \"{s3}\"")
    print(f"Test 3 Output: {result3}") # Expected: 2
````

````python
# round down appraoch
from collections import Counter
def longestPalindrome(s: str) -> int:
    table = Counter(s)
    result = 0
    for count in table.values():
        count = count // 2 * 2
        result += count
    if len(s) > result:
        result += 1
    return result

if __name__ == '__main__':
    # Test Case 1
    s1 = "abccccdd"
    result1 = longestPalindrome(s1)
    print(f"Test 1 Input: s = \"{s1}\"")
    print(f"Test 1 Output: {result1}") # Expected: 7

    # Test Case 2
    s2 = "a"
    result2 = longestPalindrome(s2)
    print(f"\nTest 2 Input: s = \"{s2}\"")
    print(f"Test 2 Output: {result2}") # Expected: 1

    # Test Case 3 (All same characters)
    s3 = "bb"
    result3 = longestPalindrome(s3)
    print(f"\nTest 3 Input: s = \"{s3}\"")
    print(f"Test 3 Output: {result3}") # Expected: 2
````

````python
# set approach
def longestPalindrome(s: str) -> int:
    remained = set()
    for letter in s:
        if letter not in remained:
            remained.add(letter)
        else:
            remained.remove(letter)
    if remained:
        return len(s) - len(remained) + 1
    else:
        return len(s)

if __name__ == '__main__':
    # Test Case 1
    s1 = "abccccdd"
    result1 = longestPalindrome(s1)
    print(f"Test 1 Input: s = \"{s1}\"")
    print(f"Test 1 Output: {result1}") # Expected: 7

    # Test Case 2
    s2 = "a"
    result2 = longestPalindrome(s2)
    print(f"\nTest 2 Input: s = \"{s2}\"")
    print(f"Test 2 Output: {result2}") # Expected: 1

    # Test Case 3 (All same characters)
    s3 = "bb"
    result3 = longestPalindrome(s3)
    print(f"\nTest 3 Input: s = \"{s3}\"")
    print(f"Test 3 Output: {result3}") # Expected: 2
````
