# 278. First Bad Version

[Link to Problem](https://leetcode.com/problems/first-bad-version/)

### Description

You are a product manager and currently leading a team to develop a new product. Unfortunately, the latest version of your product fails the quality check. Since each version is developed based on the previous version, all the versions after a bad version are also bad.

Suppose you have `n` versions `[1, 2, ..., n]` and you want to find out the first bad one, which causes all the following ones to be bad.

You are given an API `bool isBadVersion(version)` which returns whether `version` is bad. Implement a function to find the first bad version. You should minimize the number of calls to the API.

**Example 1:**
```
Input: n = 5, bad = 4
Output: 4
Explanation:
call isBadVersion(3) -> false
call isBadVersion(5) -> true
call isBadVersion(4) -> true
Then 4 is the first bad version.
```

**Example 2:**
```
Input: n = 1, bad = 1
Output: 1
```

**Constraints:**
- `1 <= bad <= n <= 2^31 - 1`

## My Intuition

_Write down your thoughts, ideas, and insights here._

- **Observations:**
  1. Use binary search
  2. if middle is false then, left = middle + 1
  3. else right = middle - 1
- **Expected approach and complexity:**
  1. Time:  O(logn)
  2. Space: O(1)

````python
# Best practice in read word (use first < last) to know the stop will be on first = last
class Solution:
    def firstBadVersion(self, n: int) -> int:
        first, last = 1, n

        while first < last:
            mid = first + (last - first) // 2

            if isBadVersion(mid):
                last = mid
            else:
                first = mid + 1

        return first
````

````python
# The isBadVersion API is already defined for you.
# def isBadVersion(version: int) -> bool:

def firstBadVersion(n: int) -> int:
    left = 1
    right = n
    while left <= right:
        middle = (left + right) // 2
        if isBadVersion(middle):
            right = middle - 1
        else:
            left = middle + 1
    return left

if __name__ == '__main__':
    # Helper setup to mock the LeetCode API environment for local testing
    def run_test(n, bad_version):
        global isBadVersion
        isBadVersion = lambda v: v >= bad_version
        return firstBadVersion(n)

    # Test Case 1
    n1, bad1 = 5, 4
    print(f"Test 1 Input: n = {n1}, bad = {bad1}")
    print(f"Test 1 Output: {run_test(n1, bad1)}") # Expected: 4

    # Test Case 2
    n2, bad2 = 1, 1
    print(f"\nTest 2 Input: n = {n2}, bad = {bad2}")
    print(f"Test 2 Output: {run_test(n2, bad2)}") # Expected: 1
````
