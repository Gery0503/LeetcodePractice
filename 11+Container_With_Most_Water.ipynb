# 11. Container With Most Water

[Link to Problem](https://leetcode.com/problems/container-with-most-water/)

### Description

You are given an integer array `height` of length `n`. There are `n` vertical lines drawn such that the two endpoints of the `i`th line are `(i, 0)` and `(i, height[i])`.

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return _the maximum amount of water a container can store_.

**Notice** that you may not slant the container.

**Example 1:**
```
Input: height = [1,8,6,2,5,4,8,3,7]
Output: 49
Explanation: The above vertical lines are represented by array [1,8,6,2,5,4,8,3,7]. In this case, the max area of water (blue section) the container can contain is 49.
```

**Example 2:**
```
Input: height = [1,1]
Output: 1
```

**Constraints:**
- `n == height.length`
- `2 <= n <= 10^5`
- `0 <= height[i] <= 10^4`

- **Observations:**
  1. Use two-pointers approach: create left and right variables
  2. Keep current maximize volume water
  3. If left is higher than right, then right point goes left until it reach the one higher then previous right (which calculate before)
  4. Compare current volume with maximize volume

````python
# Readable appraoch
from typing import List

def maxArea(height: List[int]) -> int:
    left = 0
    right = len(height)-1
    max_amount = 0
    while left < right:
        cur_amount = (right - left) * min(height[left], height[right])
        max_amount = max(cur_amount, max_amount)
        if height[left] >= height[right]:
            right -= 1
        else:
            left += 1

    return max_amount

if __name__ == '__main__':
    # Test Case 1
    height1 = [1, 8, 6, 2, 5, 4, 8, 3, 7]
    result1 = maxArea(height1)
    print(f"Test 1 Input: height = {height1}")
    print(f"Test 1 Output: {result1}") # Expected: 49

    # Test Case 2
    height2 = [1, 1]
    result2 = maxArea(height2)
    print(f"\nTest 2 Input: height = {height2}")
    print(f"Test 2 Output: {result2}") # Expected: 1

    # Test Case 3
    height3 = [2,3,4,5,18,17,6]
    result3 = maxArea(height3)
    print(f"\nTest 3 Input: height = {height3}")
    print(f"Test 3 Output: {result3}") # Expected: 17
````

````python
from typing import List

def maxArea(height: List[int]) -> int:
    l_hei = height[0]
    r_hei = height[-1]
    max_amount = (len(height)-1) * min(l_hei, r_hei)
    is_left_higher = l_hei > r_hei
    
    left = 0
    right = len(height)-1
    while left < right:
        if is_left_higher:
            while height[right] <= r_hei and left < right:
                right -= 1
            r_hei = height[right]
        else:
            while height[left] <= l_hei and left < right:
                left += 1
            l_hei = height[left]
                
        cur_amount = (right - left) * min(height[left], height[right])
        max_amount = max(cur_amount, max_amount)
        is_left_higher = height[left] > height[right]
        
    return max_amount

if __name__ == '__main__':
    # Test Case 1
    height1 = [1, 8, 6, 2, 5, 4, 8, 3, 7]
    result1 = maxArea(height1)
    print(f"Test 1 Input: height = {height1}")
    print(f"Test 1 Output: {result1}") # Expected: 49

    # Test Case 2
    height2 = [1, 1]
    result2 = maxArea(height2)
    print(f"\nTest 2 Input: height = {height2}")
    print(f"Test 2 Output: {result2}") # Expected: 1

    # Test Case 3
    height3 = [2,3,4,5,18,17,6]
    result3 = maxArea(height3)
    print(f"\nTest 3 Input: height = {height3}")
    print(f"Test 3 Output: {result3}") # Expected: 17
````
