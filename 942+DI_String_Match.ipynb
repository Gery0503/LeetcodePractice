### Description

Given a string `s` of length `n` consisting only of the characters `'I'` and `'D'`, return any permutation of the integers `[0, 1, ..., n]` such that:

- If `s[i] == 'I'`, then `perm[i] < perm[i + 1]`.
- If `s[i] == 'D'`, then `perm[i] > perm[i + 1]`.

**Example 1:**
```
Input: s = "IDID"
Output: [0,4,1,3,2]
```

**Example 2:**
```
Input: s = "III"
Output: [0,1,2,3]
```

**Example 3:**
```
Input: s = "DDI"
Output: [3,2,0,1]
```

**Constraints:**
- `1 <= s.length <= 10^5`
- `s[i]` is either `'I'` or `'D'`

# âœ… **Overall Feedback**

Your logic is *almost* correct, but the implementation has a **fundamental flaw**:

### âŒ **You should NOT append based on the previous appended number.**

The DI String Match problem is solved using **two pointers**:

* a `low` pointer (starts at 0)
* a `high` pointer (starts at n)

At each character:

* If `'I'`: append `low`, then low++
* If `'D'`: append `high`, then high--

And **the final number MUST be exactly `low` (== high)`**, no need to inspect the last char or do manual adjustment.

Your logic instead tries to compute the last element by looking at the last appended number, which breaks when the result array diverges from pointer states.

# â­ Recommended Clean Version (still O(n), same style as yours)

```python
def diStringMatch(s):
    low, high = 0, len(s)
    result = []

    for char in s:
        if char == "I":
            result.append(low)
            low += 1
        else:  # char == "D"
            result.append(high)
            high -= 1

    # low == high at this point
    result.append(low)
    return result
```

### Why this is correct?

Because:

* For every `I`, we assign the smallest unused number.
* For every `D`, we assign the largest unused number.
* After processing all chars, only one number remains â†’ `low == high`.
