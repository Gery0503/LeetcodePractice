# 345. Reverse Vowels of a String

[Link to Problem](https://leetcode.com/problems/reverse-vowels-of-a-string/)

### Description

Given a string `s`, reverse only all the vowels in the string and return it.

The vowels are `'a'`, `'e'`, `'i'`, `'o'`, and `'u'`, and they can appear in both lower and upper cases, more than once.

**Example 1:**
```
Input: s = "IceCreAm"
Output: "AceCreIm"
Explanation: 
The vowels are ['I', 'e', 'e', 'A']. On reversing the vowels, s becomes "AceCreIm".
```

**Example 2:**
```
Input: s = "leetcode"
Output: "leotcede"
```

**Constraints:**
- `1 <= s.length <= 3 * 10^5`
- `s` consist of printable ASCII characters.

## My Intuition

_Write down your thoughts, ideas, and insights here._

- **Observations:**
  1. Two pointer method: search from both left-to-right and right-to-left at the same time.
  2. Could be done in one pass.
- **Edge cases:**
  - (No edge cases)
- **Expected approach and complexity:**
  1. Time: O(1)
  2. Space: O(n)

````python
def reverseVowels(s: str) -> str:
    vowels = {"a", "e", "i", "o", "u", "A", "E", "I", "O", "U"}
    left_half = ""
    right_half = ""
    left, right = 0, len(s)-1
    while left < right:
        if s[left] in vowels and s[right] in vowels:
            left_half = left_half + s[right]
            right_half = s[left] + right_half
            left += 1
            right -= 1
            
        elif s[left] in vowels:
            right_half = s[right] + right_half
            right -= 1
            
        elif s[right] in vowels:
            left_half = left_half + s[left]
            left += 1
            
        else:
            left_half = left_half + s[left]
            right_half = s[right] + right_half
            left += 1
            right -= 1
            
    if left == right:
        return left_half + s[left] + right_half
    else:  
        return left_half + right_half

if __name__ == '__main__':
    # Test Case 1
    s1 = "IceCreAm"
    result1 = reverseVowels(s1)
    print(f"Test 1 Input: s = \"{s1}\"")
    print(f"Test 1 Output: \"{result1}\"") # Expected: "AceCreIm"

    # Test Case 2
    s2 = "leetcode"
    result2 = reverseVowels(s2)
    print(f"Test 2 Input: s = \"{s2}\"")
    print(f"Test 2 Output: \"{result2}\"") # Expected: "leotcede"

    # Test Case 3
    s3 = " "
    result3 = reverseVowels(s3)
    print(f"Test 3 Input: s = \"{s3}\"")
    print(f"Test 3 Output: \"{result3}\"") # Expected: " "
````
