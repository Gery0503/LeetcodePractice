# 290. Word Pattern

[Link to Problem](https://leetcode.com/problems/word-pattern/)

### Description

Given a `pattern` and a string `s`, find if `s` follows the same pattern.

Here **follow** means a full match, such that there is a bijection between a letter in `pattern` and a **non-empty** word in `s`. Specifically:
- Each letter in `pattern` maps to **exactly** one unique word in `s`.
- Each unique word in `s` maps to **exactly** one letter in `pattern`.
- No two letters map to the same word, and no two words map to the same letter.

**Example 1:**
```
Input: pattern = "abba", s = "dog cat cat dog"
Output: true
```

**Example 2:**
```
Input: pattern = "abba", s = "dog cat cat fish"
Output: false
```

**Example 3:**
```
Input: pattern = "aaaa", s = "dog cat cat dog"
Output: false
```

**Constraints:**
- `1 <= pattern.length <= 300`
- `pattern` contains only lower-case English letters.
- `1 <= s.length <= 3000`
- `s` contains only lowercase English letters and spaces `' '`.
- `s` **does not contain** any leading or trailing spaces.
- All the words in `s` are separated by a **single space**.

## My Intuition

_Write down your thoughts, ideas, and insights here._

- **Observations:**
  1. A bijection is both onto and one-to-one
  2. Create two dictionary to do bijection checking. One is for letter in pattern to word in s and another one is for word in s to letter in pattern.
- **Expected approach and complexity:**
  1. Time:  O(n)
  2. Space: O(n)

````python
def wordPattern(pattern: str, s: str) -> bool:
    words = s.split(' ')
    if len(pattern) != len(words):
        return False

    pattern_dict = {}
    s_dict = {}
    for let, word in zip(pattern, words):
        if let not in pattern_dict and word not in s_dict:
            pattern_dict[let] = word
            s_dict[word] = let
        elif pattern_dict.get(let, '') != word or s_dict.get(word, '') != let:
            return False
    return True
               

if __name__ == '__main__':
    # Test Case 1
    pattern1 = "abba"
    s1 = "dog cat cat dog"
    result1 = wordPattern(pattern1, s1)
    print(f"Test 1 Input: pattern = \"{pattern1}\", s = \"{s1}\"")
    print(f"Test 1 Output: {result1}") # Expected: True

    # Test Case 2
    pattern2 = "abba"
    s2 = "dog cat cat fish"
    result2 = wordPattern(pattern2, s2)
    print(f"\nTest 2 Input: pattern = \"{pattern2}\", s = \"{s2}\"")
    print(f"Test 2 Output: {result2}") # Expected: False

    # Test Case 3
    pattern3 = "aaaa"
    s3 = "dog cat cat dog"
    result3 = wordPattern(pattern3, s3)
    print(f"\nTest 3 Input: pattern = \"{pattern3}\", s = \"{s3}\"")
    print(f"Test 3 Output: {result3}") # Expected: False
````

````python
def wordPattern(pattern: str, s: str) -> bool:
    words = s.split(' ')
    if len(pattern) != len(words):
        return False
    return len(set(pattern)) == len(set(words)) == len(set(zip(pattern, words)))
               
if __name__ == '__main__':
    # Test Case 1
    pattern1 = "aba"
    s1 = "cat cat cat dog"
    result1 = wordPattern(pattern1, s1)
    print(f"Test 1 Input: pattern = \"{pattern1}\", s = \"{s1}\"")
    print(f"Test 1 Output: {result1}") # Expected: False
````
