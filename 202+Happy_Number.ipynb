# 202. Happy Number

[Link to Problem](https://leetcode.com/problems/happy-number/)

### Description
A **happy number** is defined by the following process: Starting with any positive integer, replace the number by the **sum of the squares of its digits**. Repeat the process until the number either equals 1 (where it will stay) or loops endlessly in a cycle that does not include 1. Return `true` if it becomes 1; otherwise, return `false`.

---
**Example 1:**

Input: `n = 19`
Output: `true`

**Example 2:**

Input: `n = 2`
Output: `false`

---
**Constraints:**
- `1 <= n <= 2^31 - 1`


## My intuition
- use while loop to detect number changes
- if number no changes and the final number is not equal 1, then return false

In the end, I use history set.


````python
class Solution:
    def isHappy(self, n: int) -> bool:
        """
        Determine if n is a happy number.
        """
        history = set()
        while True:
            # define new number
            digits = n
            new_n = 0
            while digits != 0:
                new_n += (digits % 10)**2 
                digits //= 10

            if new_n in history:
                break

            n = new_n
            history.add(n)
        
        return n == 1
````

## ‚úÖ Strengths

1. **Correctness**

   * The use of a `history` set ensures loop detection, preventing infinite cycles.
   * The digit-sum-square transformation is correctly implemented with `% 10` and `// 10`.
   * Test assertions cover both happy and unhappy cases.

2. **Clarity**

   * The function is short, and the logic is easy to follow.
   * Comments inside the loop clarify what is happening.

3. **Edge Cases**

   * Handles `n = 1` directly, since it will eventually return `True`.
   * Large values are safe because each step reduces the number to at most `9¬≤ * digits` (bounded to ‚â§ 243 for any input).

3. **Efficiency (Optional Optimization)**

   * Instead of a `set` for loop detection, one could use **Floyd‚Äôs Cycle Detection (Tortoise and Hare)**.
   * This avoids `O(k)` space and uses `O(1)` space.

````python
def isHappy(self, n: int) -> bool:
       def next_number(x):
           return sum(int(d)**2 for d in str(x))
       
       slow, fast = n, next_number(n)
       while fast != 1 and slow != fast:
           slow = next_number(slow)
           fast = next_number(next_number(fast))
       return fast == 1
````

## üîç Suggested Final Code (Readability + Correctness)

```python
class Solution:
    def isHappy(self, n: int) -> bool:
        """
        Determine if a number is a happy number.
        
        Args:
            n (int): The input positive integer.
        
        Returns:
            bool: True if n is a happy number, False otherwise.
        """
        history = set()
        while n != 1 and n not in history:
            history.add(n)
            n = sum(int(d)**2 for d in str(n))
        return n == 1
```

What is the time complexity?

Good question üëç Let‚Äôs analyze the **time complexity** of the *Happy Number* problem carefully.

---

## Step 1: Range of Values After Transformation

* For a number `n` with `k` digits, the maximum sum of squares is:
  [
  k \cdot 9^2 = 81k
  ]
* For example:

  * `n = 2^31 - 1 ‚âà 2.1 \times 10^9` has 10 digits ‚Üí max sum of squares = ( 10 \cdot 81 = 810 ).
* That means no matter how large `n` is, after **at most one iteration**, it is bounded by 810.
* From then on, all transformations stay within the range `[1, 810]`.

---

## Step 2: Loop Behavior

* There are at most **810 possible states**.
* Once we revisit a number, the cycle is detected (either reaches 1 or loops forever).
* So the loop runs **at most 810 iterations**.

---

## Step 3: Cost Per Iteration

* In each iteration, we compute the sum of squares of digits.
* Cost is **O(log n)** for digit extraction (since number has `log‚ÇÅ‚ÇÄ(n)` digits).
* But since `n` shrinks quickly and is capped at 810 after the first step, the digit count is **‚â§ 3** for the rest of the process.
* So effectively, digit extraction becomes **O(1)**.

---

## Final Complexity

* **Time Complexity:**

  * General form: **O(log n)** (from digit extraction in the very first step).
  * Tight bound: **O(1)**, since the loop is bounded by a constant (‚â§ 810) and digits are ‚â§ 3 afterward.
* **Space Complexity:**

  * With a `set` for history ‚Üí **O(810) = O(1)**.
  * With Floyd‚Äôs cycle detection ‚Üí **O(1)**.

---

**‚ÄúTime complexity is O(log n), but effectively constant time since the sequence always converges within a small bounded range (‚â§ 810). Space is O(1).‚Äù**
