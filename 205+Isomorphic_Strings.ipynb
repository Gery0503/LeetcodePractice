# 205. Isomorphic Strings

[Link to Problem](https://leetcode.com/problems/isomorphic-strings/)

### Description

Given two strings `s` and `t`, *determine if they are isomorphic*.

Two strings `s` and `t` are isomorphic if the characters in `s` can be replaced to get `t`.

All occurrences of a character must be replaced with another character while preserving the order of characters. No two characters may map to the same character, but a character may map to itself.

**Example 1:**
```
Input: s = "egg", t = "add"
Output: true
```

**Example 2:**
```
Input: s = "foo", t = "bar"
Output: false
```

**Example 3:**
```
Input: s = "paper", t = "title"
Output: true
```

**Constraints:**
- `1 <= s.length <= 5 * 10^4`
- `t.length == s.length`
- `s` and `t` consist of any valid ascii character.

- **Expected approach and complexity:**
  1. Time:  O(n)

````python
def isIsomorphic(s: str, t: str) -> bool:
    using = set()
    mapping = {}
    for char_s, char_t in zip(s, t):
        if char_s not in mapping:
            if char_t in using:
                return False
                
            mapping[char_s] = char_t
            using.add(char_t)
        else:
            if mapping.get(char_s) != char_t:
                return False

    return True
        

if __name__ == '__main__':
    # Test Case 1
    s1 = "egg"
    t1 = "add"
    result1 = isIsomorphic(s1, t1)
    print(f"Test 1 Input: s = \"{s1}\", t = \"{t1}\"")
    print(f"Test 1 Output: {result1}") # Expected: True

    # Test Case 2
    s2 = "foo"
    t2 = "bar"
    result2 = isIsomorphic(s2, t2)
    print(f"\nTest 2 Input: s = \"{s2}\", t = \"{t2}\"")
    print(f"Test 2 Output: {result2}") # Expected: False

    # Test Case 3
    s3 = "paper"
    t3 = "title"
    result3 = isIsomorphic(s3, t3)
    print(f"\nTest 3 Input: s = \"{s3}\", t = \"{t3}\"")
    print(f"Test 3 Output: {result3}") # Expected: True

    # Test Case 4
    s4 = "babc"
    t4 = "baba"
    result4 = isIsomorphic(s4, t4)
    print(f"\nTest 4 Input: s = \"{s4}\", t = \"{t4}\"")
    print(f"Test 4 Output: {result4}") # Expected: False
````

````python
# one-line Python solution
def isIsomorphic(s: str, t: str) -> bool:
    return len(set(s)) == len(set(t)) == len(set(zip(s, t)))
````

````python
# fastest solution
class Solution:
    def isIsomorphic(self, s: str, t: str) -> bool:
        # Extended ASCII goes to 255
        indexS = [0] * 256
        indexT = [0] * 256
        
        length = len(s)
        
        if length != len(t):
            return False
        
        for i in range(length):
            # Check if the index of the current character in string s is different from the index of the corresponding character in string t
            if indexS[ord(s[i])] != indexT[ord(t[i])]: 
                return False
                
            # updating position of current character
            indexS[ord(s[i])] = i + 1 
            indexT[ord(t[i])] = i + 1
        
        return True
````
