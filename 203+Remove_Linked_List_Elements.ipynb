# 203. Remove Linked List Elements

[Link to Problem](https://leetcode.com/problems/remove-linked-list-elements/)

### Description

Given the `head` of a linked list and an integer `val`, remove all the nodes of the linked list that has `Node.val == val`, and return the new head.

**Example 1:**
```
Input: head = [1,2,6,3,4,5,6], val = 6
Output: [1,2,3,4,5]
```

**Example 2:**
```
Input: head = [], val = 1
Output: []
```

**Example 3:**
```
Input: head = [7,7,7,7], val = 7
Output: []
```

**Constraints:**
- The number of nodes in the list is in the range `[0, 10^4]`.
- `1 <= Node.val <= 50`
- `0 <= val <= 50`

## My Intuition

_Write down your thoughts, ideas, and insights here._

- **Observations:**
  1. Memorize previous node.
  2. If current_node val needs to be removed, then previous_node.next = current_node.next and current_node.next = None
  3. else do nothing
- **Edge cases:**
  - First one is the val
- **Expected approach and complexity:**
  1. Time:  O(n)
  2. Space: O(1)

````python
# One node solution
class Solution:
    def removeElements(self, head, val):
        dummy_head = ListNode(-1)
        dummy_head.next = head
        
        current_node = dummy_head
        while current_node.next != None:
            if current_node.next.val == val:
                current_node.next = current_node.next.next
            else:
                current_node = current_node.next
                
        return dummy_head.next
````

````python
# The first draft solution
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeElements(head: Optional[ListNode], val: int) -> Optional[ListNode]:
    dummy = ListNode()
    dummy.next = head
        
    previous = dummy
    current = dummy.next
    while current:
        if current.val == val:
            previous.next = current.next
            current.next = None
            current = previous.next
        else:
            previous = current
            current = current.next
    return dummy.next

if __name__ == '__main__':
    # Helper function to build a linked list from an array
    def build_list(arr):
        if not arr: return None
        head = ListNode(arr[0])
        curr = head
        for v in arr[1:]:
            curr.next = ListNode(v)
            curr = curr.next
        return head

    # Helper function to print a linked list as an array
    def print_list(node):
        res = []
        while node:
            res.append(node.val)
            node = node.next
        return res

    # Test Case 1
    head1 = build_list([1, 2, 6, 3, 4, 5, 6])
    val1 = 6
    print(f"Test 1 Input: head = [1, 2, 6, 3, 4, 5, 6], val = {val1}")
    result1 = removeElements(head1, val1)
    print(f"Test 1 Output: {print_list(result1)}") # Expected: [1, 2, 3, 4, 5]

    # Test Case 2
    head2 = build_list([])
    val2 = 1
    print(f"\nTest 2 Input: head = [], val = {val2}")
    result2 = removeElements(head2, val2)
    print(f"Test 2 Output: {print_list(result2)}") # Expected: []

    # Test Case 3
    head3 = build_list([7, 7, 7, 7])
    val3 = 7
    print(f"\nTest 3 Input: head = [7, 7, 7, 7], val = {val3}")
    result3 = removeElements(head3, val3)
    print(f"Test 3 Output: {print_list(result3)}") # Expected: []
````
