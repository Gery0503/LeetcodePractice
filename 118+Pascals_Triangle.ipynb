# 118. Pascal's Triangle

[Link to Problem](https://leetcode.com/problems/pascals-triangle/)

### Description

Given an integer `numRows`, return the first `numRows` of Pascal's triangle.

In Pascal's triangle, each number is the sum of the two numbers directly above it, as shown:

```
      [1]
    [1] [1]
  [1] [2] [1]
 [1] [3] [3] [1]
[1] [4] [6] [4] [1]
```

**Example 1:**
```
Input: numRows = 5
Output: [[1],[1,1],[1,2,1],[1,3,3,1],[1,4,6,4,1]]
```

**Example 2:**
```
Input: numRows = 1
Output: [[1]]
```

**Constraints:**
- `1 <= numRows <= 30`


## My Intuition

_Write down your thoughts, ideas, and insights here._

- **Observations:**
  1. The first and last element of every row is always 1.
  2. The value of an element at index `j` in row `i` is the sum of the elements at indices `j-1` and `j` in row `i-1`.
- **Edge cases:**
  - What if numRows is 1 or 2?
- **Expected approach and complexity:**
  1. Time: O(n**2)
  2. Space: O(1)

````python
from typing import List

def generate(numRows: int) -> List[List[int]]:
    if numRows == 1:
        return [[1]]
    elif numRows == 2:
        return [[1],[1, 1]]
    row = [[1],[1, 1]]
    i = 2
    previous = [1, 1]
    while i < numRows:
        current = [1]
        for j in range(len(previous)-1):
            current.append(previous[j]+previous[j+1])
        current.append(1)
        row.append(current)
        previous = current
        i += 1
    return row
        

if __name__ == '__main__':
    # Test Case 1
    numRows1 = 5
    result1 = generate(numRows1)
    print(f"Input: {numRows1}")
    print(f"Output: {result1}") # Expected: [[1],[1,1],[1,2,1],[1,3,3,1],[1,4,6,4,1]]

    # Test Case 2
    numRows2 = 1
    result2 = generate(numRows2)
    print(f"Input: {numRows2}")
    print(f"Output: {result2}") # Expected: [[1]]
````

## My Intuition (corrected)
### Observations:
The first and last element of every row is always 1.

The value of an element at index j in row i is the sum of the elements at indices j-1 and j in row i-1.

### Edge cases:
The base case for the loop should correctly handle numRows = 1 and numRows = 2.

Expected approach and complexity:
- Time: O(N²) - Correct, as we generate $1 + 2 + ... + N$ elements.
- Space: O(N²) - Since we store all $O(N^2)$ elements in the resulting triangle.


````python
# Refactored version
from typing import List

def generate(numRows: int) -> List[List[int]]:
    """
    Generates the first numRows of Pascal's Triangle.
    
    Time Complexity: O(N^2), where N is numRows.
    Space Complexity: O(N^2), for storing the resulting triangle.
    """
    # Initialize the result list
    triangle: List[List[int]] = []
    
    # Iterate through each row we need to generate (from row 0 up to numRows - 1)
    for i in range(numRows):
        # A new row always starts and ends with 1.
        # The row has i + 1 elements.
        current_row = [1] * (i + 1)
        
        # If we are past the first two rows (i > 1), calculate the inner elements.
        if i >= 2:
            previous_row = triangle[i - 1]
            
            # The inner elements are the sum of the two elements above them (j-1 and j)
            # in the previous row.
            # We iterate from index 1 up to the second-to-last element (i - 1).
            for j in range(1, i):
                current_row[j] = previous_row[j - 1] + previous_row[j]
        
        # Add the completed row to the triangle
        triangle.append(current_row)
        
    return triangle
````
