# 941. Valid Mountain Array

[Link to Problem](https://leetcode.com/problems/valid-mountain-array/)

### Description

Given an array of integers `arr`, return `true` **if and only if** it is a valid mountain array.

Recall that `arr` is a mountain array if and only if:
- `arr.length >= 3`
- There exists some index `i` with `0 < i < arr.length - 1` such that:
  - `arr[0] < arr[1] < ‚Ä¶ < arr[i - 1] < arr[i]`
  - `arr[i] > arr[i + 1] > ‚Ä¶ > arr[arr.length - 1]`

**Example 1:**
```
Input: arr = [2, 1]
Output: false
```

**Example 2:**
```
Input: arr = [3, 5, 5]
Output: false
```

**Example 3:**
```
Input: arr = [0, 3, 2, 1]
Output: true
```

### Constraints:

- 1 <= arr.length <= 104
- 0 <= arr[i] <= 104

## My Intuition

- Observations:
  1. Increasing before decreasing
  2. Turning point should be one only.
  3. Adjacent points cannot have the same height.
- Edge cases to consider:
  1. only increasing
  2. only decreasing
  3. plateau
  4. short array
- Expected time/space complexity:
  1. Time:  O(n)
  2. Space: O(1)


````python
def validMountainArray(arr):
    if len(arr) < 3:
        return False
        
    increasing = False
    decreasing = False
    current_height = arr[0]
    for height in arr[1:]:
        if height == current_height:
            return False
        if decreasing is True and height > current_height:
            return False
        
        if height > current_height:
            increasing = True
        else:
            decreasing = True

        current_height = height
        
    return increasing and decreasing

if __name__ == '__main__':
    arr = [0, 3, 2, 1]
    print(validMountainArray(arr))  # replace with your own tests after implementation
````

Below is **the best, cleanest, and most readable solution** for **941. Valid Mountain Array**, following the optimal two-pointer walk-up / walk-down pattern.

# ‚úÖ Best Solution (Python)

```
def validMountainArray(arr):
    n = len(arr)
    if n < 3:
        return False

    i = 0

    # walk up
    while i + 1 < n and arr[i] < arr[i + 1]:
        i += 1

    # peak cannot be first or last
    if i == 0 or i == n - 1:
        return False

    # walk down
    while i + 1 < n and arr[i] > arr[i + 1]:
        i += 1

    return i == n - 1
```

---

# ‚≠ê Why This Is The Best Solution

### **1. Super clean ‚Äútwo-phase‚Äù logic**

* Phase 1: strictly increasing
* Phase 2: strictly decreasing

### **2. Ensures the peak is valid**

The peak must be **not** at index `0` or `n-1`.

### **3. One pass, O(n) time, O(1) space**

The simplest solution using only pointer movement.

### **4. Easy to reason about**

No flags. No complicated transitions. Just walk the array.

### **5. Perfectly handles all corner cases**

* `[2,1]` ‚Üí false
* `[3,5,5]` ‚Üí false
* `[3,4,2]` ‚Üí true
* `[1,2,3]` ‚Üí false
* `[3,2,1]` ‚Üí false
* `[0,3,2,1]` ‚Üí true

# üéØ Quick Intuition

A valid mountain looks like:

```

     /  \
```

So we:

1. Walk **up** until the slope stops rising
2. The index we stop at must be a **peak**
3. Then walk **down**
4. If we reach the end, it‚Äôs a valid mountain
