# 83. Remove Duplicates from Sorted List

[Link to Problem](https://leetcode.com/problems/remove-duplicates-from-sorted-list/)

### Description

Given the `head` of a sorted linked list, delete all duplicates such that each element appears only once. Return the linked list **sorted** as well.

**Example 1:**
```
Input: head = [1,1,2]
Output: [1,2]
```

**Example 2:**
```
Input: head = [1,1,2,3,3]
Output: [1,2,3]
```

**Constraints:**
- The number of nodes in the list is in the range `[0, 300]`.
- `-100 <= Node.val <= 100`
- The list is guaranteed to be **sorted** in ascending order.

## My Intuition

_Write down your thoughts, ideas, and insights here._

- **Observations:**
  1. Create current node variable to compare with the next node
  2. If next node's value is not the same, then override current node with next node
  3. else move on to next next node
- **Expected approach and complexity:**
  1. Time:  O(n)
  2. Space: O(1)

````python
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def deleteDuplicates(head: Optional[ListNode]) -> Optional[ListNode]:
    if head is None:
        return None
        
    cur_node = head
    while cur_node.next:
        if cur_node.val != cur_node.next.val:
            cur_node = cur_node.next
        else:
            cur_node.next = cur_node.next.next
            
    return head
    
if __name__ == '__main__':
    # Helper function to build a linked list from an array
    def build_list(arr):
        if not arr: return None
        head = ListNode(arr[0])
        curr = head
        for val in arr[1:]:
            curr.next = ListNode(val)
            curr = curr.next
        return head

    # Helper function to print a linked list as an array
    def print_list(node):
        res = []
        while node:
            res.append(node.val)
            node = node.next
        return res

    # Test Case 1
    head1 = build_list([1, 1, 2])
    print(f"Test 1 Input: head = [1, 1, 2]")
    result1 = deleteDuplicates(head1)
    print(f"Test 1 Output: {print_list(result1)}") # Expected: [1, 2]

    # Test Case 2
    head2 = build_list([1, 1, 2, 3, 3])
    print(f"\nTest 2 Input: head = [1, 1, 2, 3, 3]")
    result2 = deleteDuplicates(head2)
    print(f"Test 2 Output: {print_list(result2)}") # Expected: [1, 2, 3]

    # Test Case 3 (Empty list)
    head3 = build_list([])
    print(f"\nTest 3 Input: head = []")
    result3 = deleteDuplicates(head3)
    print(f"Test 3 Output: {print_list(result3)}") # Expected: []
````
