# 696. Count Binary Substrings

[Link to Problem](https://leetcode.com/problems/count-binary-substrings/description/)

### Description
Given a binary string `s`, return the number of non-empty substrings that have the **same number of 0's and 1's**, and where all the 0's and all the 1's in the substring are **grouped consecutively**.
A substring is a contiguous sequence of characters within the string.

---
**Example 1:**

Input: `s = "00110011"`
Output: `6`
Explanation: The valid substrings are "0011", "01", "1100", "10", "0011", and "01".

**Example 2:**

Input: `s = "10101"`
Output: `4`
Explanation: The valid substrings are "10", "01", "10", and "01".

---
**Constraints:**
- `1 <= s.length <= 50,000`
- `s` consists only of the characters `'0'` and `'1'`.


### My intuition
 - count how many cuts for groups
 - count consecutive for each group

```python
class Solution:
    def countBinarySubstrings(self, s: str) -> int:
        s0 = 0
        s1 = 0
        counts = 0
        prev = s[0]
        for num in s:
            if num != prev:
                # reset
                if num == '0':
                    s0 = 0
                else:
                    s1 = 0
            if num == '0':
                s0 += 1
                if s1 > 0:
                    s1 -= 1
                    counts += 1
            else:
                s1 += 1
                if s0 > 0:
                    s0 -= 1
                    counts += 1
            prev = num
        return counts
# Time: O(n)
# Space: O(1)
```

### 2. **Logic Style**

Your approach is clever but harder to follow. It uses two counters (`s0`, `s1`) that shrink as the opposite grows.
üëâ A more intuitive way is the **group length method**:

* Count lengths of consecutive groups.
* The answer is the sum of `min(prev_group, curr_group)`.

Example refactor:

```python
class Solution:
    def countBinarySubstrings(self, s: str) -> int:
        prev_len, curr_len = 0, 1
        result = 0
        for i in range(1, len(s)):
            if s[i] == s[i-1]:
                curr_len += 1
            else:
                result += min(prev_len, curr_len)
                prev_len, curr_len = curr_len, 1
        return result + min(prev_len, curr_len)
# Time: O(n)
# Space: O(1)
```

* ‚úÖ Still `O(n)` time, `O(1)` space
* ‚úÖ Matches the typical greedy solution
* ‚úÖ Easier for others to read

### Why did your version run faster?

1. **Micro-benchmark randomness (LeetCode factor)**

   * LeetCode runtime is **not deterministic**.
   * Each run is affected by server load, caching, and Python interpreter warm-up.
   * A 10‚Äì20ms gap is often just noise, not a guaranteed performance difference.

2. **Implementation differences**

   * Your `s0/s1 balance` version avoids `min()` and multiple variable swaps.
   * It only uses **two counters** (`s0`, `s1`) and updates them with simple arithmetic.
   * The greedy version does `min(prev, curr)` and frequent assignments, which adds tiny overhead.

3. **Branching behavior**

   * In your version, updates happen in a tighter loop with fewer resets.
   * Greedy version does an extra branch whenever the character switches.

4. **Memory locality**

   * Both are `O(1)` space, but your counters may stay hot in CPU registers, while greedy‚Äôs `prev_len`/`curr_len` swaps might spill to memory more often.
