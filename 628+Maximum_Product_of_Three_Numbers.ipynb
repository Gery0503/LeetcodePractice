# 628. Maximum Product of Three Numbers

[Link to Problem](https://leetcode.com/problems/maximum-product-of-three-numbers/)

### Description

Given an integer array `nums`, find three numbers whose product is maximum and return the maximum product.

**Example 1:**
```
Input: nums = [1, 2, 3]
Output: 6
```

**Example 2:**
```
Input: nums = [1, 2, 3, 4]
Output: 24
```

**Example 3:**
```
Input: nums = [-1, -2, -3]
Output: -6
```

**Constraints:**
- `3 <= nums.length <= 10^4`
- `-1000 <= nums[i] <= 1000`


```python
from typing import List

class Solution:
    def maximumProduct(self, nums: List[int]) -> int:
        # We only need the three largest numbers and the two smallest numbers
        # to cover cases with negative numbers.
        max1 = max2 = max3 = float('-inf')
        min1 = min2 = float('inf')
        
        for n in nums:
            # Update largest values
            if n > max1:
                max3, max2, max1 = max2, max1, n
            elif n > max2:
                max3, max2 = max2, n
            elif n > max3:
                max3 = n
            
            # Update smallest values
            if n < min1:
                min2, min1 = min1, n
            elif n < min2:
                min2 = n
        
        # The maximum product is either:
        # - product of the three largest values, or
        # - product of the two smallest (i.e. possibly large negatives) and the largest value.
        return max(max1 * max2 * max3, min1 * min2 * max1)
```

### ✅ Why this works

* We track the **top 3 largest** numbers (`max1, max2, max3`) and the **bottom 2 smallest** numbers (`min1, min2`) in one pass — O(n) time, O(1) extra space. ([massivealgorithms.blogspot.com][1])
* Because negative numbers can flip the sign, the maximum product might come from either:

  * the three largest (all positives, or mixed signs), or
  * two most negative + one largest positive — that's why we consider both `max1*max2*max3` and `min1*min2*max1`. ([LeetCodee][2])

This approach handles all edge cases (all positives, all negatives, mixed, zeros) correctly.
