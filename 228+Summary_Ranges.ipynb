# 228. Summary Ranges

[Link to Problem](https://leetcode.com/problems/summary-ranges/)

### Description

You are given a **sorted unique** integer array `nums`.

A **range** `[a,b]` is the set of all integers from `a` to `b` (inclusive).

Return _the **smallest sorted** list of ranges that **cover all the numbers in the array exactly**_. That is, each element of `nums` is covered by exactly one of the ranges, and there is no integer `x` such that `x` is in one of the ranges but not in `nums`.

Each range `[a,b]` in the list should be output as:
- `"a->b"` if `a != b`
- `"a"` if `a == b`

**Example 1:**
```
Input: nums = [0,1,2,4,5,7]
Output: ["0->2","4->5","7"]
Explanation: The ranges are:
[0,2] --> "0->2"
[4,5] --> "4->5"
[7,7] --> "7"
```

**Example 2:**
```
Input: nums = [0,2,3,4,6,8,9]
Output: ["0","2->4","6","8->9"]
Explanation: The ranges are:
[0,0] --> "0"
[2,4] --> "2->4"
[6,6] --> "6"
[8,9] --> "8->9"
```

**Constraints:**
- `0 <= nums.length <= 20`
- `-2^31 <= nums[i] <= 2^31 - 1`
- All the values of `nums` are **unique**.
- `nums` is sorted in ascending order.

## My Intuition

_Write down your thoughts, ideas, and insights here._

- **Observations:**
  1. Use sliding window to get each valid range
  2. Create left as range start, and create dynamic right to compare with element of nums
  3. If (right + 1) is not equal to current num, then we get one range (left, right)
  4. And we treat current num as new a

- **Expected approach and complexity:**
  1. Time:  O(n)
  2. Space: O(1)  # due to nums.length <= 20

````python
# Better solution
from typing import List

def summaryRanges(nums: List[int]) -> List[str]:
    result = []
    i = 0
    n = len(nums)
    
    while i < n:
        start = nums[i]
        
        # Fast-forward 'i' as long as the next number is consecutive
        while i + 1 < n and nums[i] + 1 == nums[i + 1]:
            i += 1
            
        if start == nums[i]:
            result.append(str(start))
        else:
            result.append(f"{start}->{nums[i]}")
            
        i += 1
    return result
````

````python
from typing import List

def summaryRanges(nums: List[int]) -> List[str]:
    if len(nums) == 0:
        return []
        
    result = []
    left, right = nums[0], nums[0]
    nums.append(nums[0])
    for i in range(1, len(nums)):
        if right + 1 == nums[i]:
            right = nums[i]
        else: 
            if left == right:
                result.append(str(left))
            else:
                result.append(f'{left}->{right}')
            left, right = nums[i], nums[i]

    return result
        

if __name__ == '__main__':
    # Test Case 1
    nums1 = [0, 1, 2, 4, 5, 7]
    result1 = summaryRanges(nums1)
    print(f"Test 1 Input: nums = {nums1}")
    print(f"Test 1 Output: {result1}") # Expected: ['0->2', '4->5', '7']

    # Test Case 2
    nums2 = [0, 2, 3, 4, 6, 8, 9]
    result2 = summaryRanges(nums2)
    print(f"\nTest 2 Input: nums = {nums2}")
    print(f"Test 2 Output: {result2}") # Expected: ['0', '2->4', '6', '8->9']

    # Test Case 3 (Empty array)
    nums3 = []
    result3 = summaryRanges(nums3)
    print(f"\nTest 3 Input: nums = {nums3}")
    print(f"Test 3 Output: {result3}") # Expected: []
````
