# 234. Palindrome Linked List

[Link to Problem](https://leetcode.com/problems/palindrome-linked-list/)

### Description

Given the `head` of a singly linked list, return `true` if it is a palindrome or `false` otherwise.

**Example 1:**
```
Input: head = [1,2,2,1]
Output: true
```

**Example 2:**
```
Input: head = [1,2]
Output: false
```

**Constraints:**
- The number of nodes in the list is in the range `[1, 10^5]`.
- `0 <= Node.val <= 9`

**Follow up:** Could you do it in `O(n)` time and `O(1)` space?

  1. Could use stack to handle Palindrome, however not this time. Because we need to do it in O(1).
  2. use slow and fast pointers to get middel node
  3. reverse right half linked list while slow pointer go ahead
- **Expected approach and complexity:**
  1. Time:  O(n)

````python
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
# Better one
class Solution:
    def reverse(self, head: ListNode) -> ListNode:
        prev = None
        curr = head
        while curr:
            next_temp = curr.next
            curr.next = prev
            prev = curr
            curr = next_temp
        return prev

    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        if not head or not head.next:
            return True

        # 1. 尋找中間節點
        slow = head
        fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        # 2. 反轉後半段，並儲存反轉後的開頭以便後續復原
        rev = self.reverse(slow)
        reversed_head_anchor = rev  
        
        # 3. 比對前半段與反轉後的後半段
        curr_head = head
        is_pal = True
        
        while rev:
            if curr_head.val != rev.val:
                is_pal = False
                break  # 發現不對稱，提早結束比對
            curr_head = curr_head.next
            rev = rev.next
            
        # 4. 復原鏈結串列 (將後半段再次反轉回來)
        self.reverse(reversed_head_anchor)
        
        return is_pal
````
