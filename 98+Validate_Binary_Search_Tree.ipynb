# 98. Validate Binary Search Tree

[Link to Problem](https://leetcode.com/problems/validate-binary-search-tree/description/)

### Description
Given the `root` of a binary tree, determine if it is a valid binary search tree (BST).

A valid BST is defined as follows:
- The left subtree of a node contains only nodes with keys **less than** the node's key.
- The right subtree of a node contains only nodes with keys **greater than** the node's key.
- Both the left and right subtrees must also be binary search trees.

---
**Example 1:**

Input: `root = [2,1,3]`
Output: `true`

**Example 2:**

Input: `root = [5,1,4,null,null,3,6]`
Output: `false`
Explanation: The root node's value is 5 but its right child's value is 4.

---
**Constraints:**
- The number of nodes in the tree is in the range `[1, 10^4]`
- `-2^31 <= Node.val <= 2^31 - 1`

My intuition: pass lower bond and upper bond

````python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# recursive DFS (Top-Down)
class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        def dfs(node: TreeNode, lower: int, upper: int) -> bool:
            if not node:  # Base case
                return True
            if not (lower < node.val < upper):
                return False
            return dfs(node.left, lower, node.val) and dfs(node.right, node.val, upper)
        return dfs(root, float('-inf'), float('inf'))
# Time: O(n)
# Space: O(h)
````

````python
# iterative DFS
class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        stack = [(root, float('-inf'), float('inf'))]
        while stack:
            node, lower, upper = stack.pop()
            if not (lower < node.val < upper):
                return False
            if node.left:
                stack.append((node.left, lower, node.val))
            if node.right:
                stack.append((node.right, node.val, upper))
        return True
# Time: O(n)
# Space: O(h)
````

Example:

````python
class Solution:
    def isValidBST(self, root: TreeNode) -> bool:
        stack, prev = [], None
        while stack or root:
            while root:
                stack.append(root)
                root = root.left
            root = stack.pop()
            if prev is not None and root.val <= prev:
                return False
            prev = root.val
            root = root.right
        return True
# Time: O(n)
# Space: O(h)
````

This strategy leverages the **sorted property of in-order traversal** for BSTs.

Here's an in-depth analysis of why the performance differs among the three `isValidBST` approaches: **recursive DFS**, **iterative DFS**, and **inorder traversal**:

---

### 🔍 1. **Recursive DFS (Top-down with bounds)**

```python
def isValidBST(root):
    def dfs(node, lower, upper):
        if not node:
            return True
        if not (lower < node.val < upper):
            return False
        return dfs(node.left, lower, node.val) and dfs(node.right, node.val, upper)
    return dfs(root, float('-inf'), float('inf'))
```

#### ✅ Pros:

* Elegant, expressive logic using bounds to verify BST rules.
* Short-circuits early if a violation is found (minimal traversal).

#### ❌ Cons:

* **Recursive overhead**: For deep trees, function call stack grows (up to O(h)).
* For large, balanced trees, recursion depth is \~log(n), but for skewed trees it's up to O(n).

#### 🧠 Performance Notes:

* Performance depends on early termination. If the BST is valid, all nodes are visited.
* Fastest for small to medium trees or invalid trees with early failures.

---

### 🔁 2. **Iterative DFS with stack and bounds**

```python
def isValidBST(root):
    stack = [(root, float('-inf'), float('inf'))]
    while stack:
        node, lower, upper = stack.pop()
        if not (lower < node.val < upper):
            return False
        if node.right:
            stack.append((node.right, node.val, upper))
        if node.left:
            stack.append((node.left, lower, node.val))
    return True
```

#### ✅ Pros:

* Avoids recursion depth issues; more robust for deep trees.
* Clear control over execution stack (useful in systems programming or constrained environments).

#### ❌ Cons:

* Slightly more bookkeeping (tuple unpacking, stack manipulation).
* Slightly more overhead due to manual stack management.

#### 🧠 Performance Notes:

* Similar number of operations to recursive DFS, but Python stack operations (list.pop/append) add overhead compared to recursive call return frames.

---

### 📜 3. **Inorder Traversal (Check if sorted)**

```python
def isValidBST(root):
    stack, inorder = [], float('-inf')
    while stack or root:
        while root:
            stack.append(root)
            root = root.left
        root = stack.pop()
        if root.val <= inorder:
            return False
        inorder = root.val
        root = root.right
    return True
```

#### ✅ Pros:

* Leverages BST property: inorder traversal yields sorted values.
* Requires only one value check (`inorder < root.val`).
* **Most cache-friendly** due to linear memory access and fewer condition checks.

#### ❌ Cons:

* Slightly less intuitive to verify correctness for non-BST trees.
* May need to visit all nodes even if a violation happens late.

#### 🧠 Performance Notes:

* Often the **fastest** for large valid trees because:

  * No bound-passing.
  * Tight inner loop with minimal comparisons.
  * Data locality is better, and fewer function calls or tuple objects are involved.
