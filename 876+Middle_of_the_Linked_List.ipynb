# 876. Middle of the Linked List

[Link to Problem](https://leetcode.com/problems/middle-of-the-linked-list/)

### Description

Given the `head` of a singly linked list, return the middle node of the linked list.

If there are two middle nodes, return **the second middle node**.

**Example 1:**
```
Input: head = [1,2,3,4,5]
Output: [3,4,5]
Explanation: The middle node of the list is node 3.
```

**Example 2:**
```
Input: head = [1,2,3,4,5,6]
Output: [4,5,6]
Explanation: Since the list has two middle nodes with values 3 and 4, we return the second one.
```

**Constraints:**
- The number of nodes in the list is in the range `[1, 100]`.
- `1 <= Node.val <= 100`

## My Intuition

- **Observations:**
  1. use two pointers

- **Expected approach and complexity:**
  1. Time:  O(n)
  2. Space: O(1)

````python
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def middleNode(head: Optional[ListNode]) -> Optional[ListNode]:
    # Implement your solution here
    slow_node = head
    fast_node = head
    while fast_node and fast_node.next:
        slow_node = slow_node.next
        fast_node = fast_node.next.next
    return slow_node

if __name__ == '__main__':
    # Helper function to build a linked list from an array
    def build_list(arr):
        if not arr: return None
        head = ListNode(arr[0])
        curr = head
        for val in arr[1:]:
            curr.next = ListNode(val)
            curr = curr.next
        return head

    # Helper function to print a linked list as an array
    def print_list(node):
        res = []
        while node:
            res.append(node.val)
            node = node.next
        return res

    # Test Case 1
    head1 = build_list([1, 2, 3, 4, 5])
    result1 = middleNode(head1)
    print(f"Test 1 Output: {print_list(result1)}") # Expected: [3, 4, 5]

    # Test Case 2
    head2 = build_list([1, 2, 3, 4, 5, 6])
    result2 = middleNode(head2)
    print(f"Test 2 Output: {print_list(result2)}") # Expected: [4, 5, 6]

    # Test Case 3
    head3 = build_list([1])
    result3 = middleNode(head3)
    print(f"Test 3 Output: {print_list(result3)}") # Expected: [1]
````
