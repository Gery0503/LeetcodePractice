### Description

Given an integer array `nums` and an integer `val`, remove all occurrences of `val` in `nums` [**in-place**](https://en.wikipedia.org/wiki/In-place_algorithm). The order of the elements may be changed. Then return _the number of elements in_ `nums` _which are not equal to_ `val`.

Consider the number of elements in `nums` which are not equal to `val` be `k`, to get accepted, you need to do the following things:
- Change the array `nums` such that the first `k` elements of `nums` contain the elements which are not equal to `val`. The remaining elements of `nums` are not important as well as the size of `nums`.
- Return `k`.

## My Intuition

_Write down your thoughts, ideas, and insights here._

- **Observations:**
  1. two-pinters: write right element when nums`[left]` == val and do right -= 1
  2. else left += 1
- **Expected approach and complexity:**
  1. Time:  O(n)
  2. Space: O(1) 

````python
# Left and right approach
from typing import List

def removeElement(nums: List[int], val: int) -> int:
    left, right = 0, len(nums)-1
    while left <= right:
        if nums[left] == val:
            nums[left] = nums[right]
            right -= 1
        else:
            left += 1
    return left

if __name__ == '__main__':
    # Test Case 1
    nums1 = [3, 2, 2, 3]
    val1 = 3
    print(f"Test 1 Input: nums = {nums1}, val = {val1}")
    k1 = removeElement(nums1, val1)
    print(f"Test 1 Output: k = {k1}, nums = {nums1[:k1]}") # Expected: k = 2, nums = [2, 2]

    # Test Case 2
    nums2 = [0, 1, 2, 2, 3, 0, 4, 2]
    val2 = 2
    print(f"\nTest 2 Input: nums = {nums2}, val = {val2}")
    k2 = removeElement(nums2, val2)
    print(f"Test 2 Output: k = {k2}, nums = {nums2[:k2]}") # Expected: k = 5, nums = [0, 1, 4, 0, 3] (order may vary)

    # Test Case 3 (Empty array)
    nums3 = []
    val3 = 0
    print(f"\nTest 3 Input: nums = {nums3}, val = {val3}")
    k3 = removeElement(nums3, val3)
    print(f"Test 3 Output: k = {k3}, nums = {nums3[:k3]}") # Expected: k = 0, nums = []

    # Test Case 4
    nums4 = [2]
    val4 = 3`
    print(f"\nTest 4 Input: nums = {nums4}, val = {val4}")
    k4 = removeElement(nums4, val4)
    print(f"Test 4 Output: k = {k4}, nums = {nums4[:k4]}") # Expected: k = 1, nums = [2]
````

````python
# fast and slow approach
from typing import List

class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        k = 0

        for i in range(len(nums)):
            if nums[i] != val:
                nums[k] = nums[i]
                k += 1
        
        return k
````

This comparison perfectly highlights the main trade-off between these two approaches:
* **Fast/Slow Pointers:** Highly efficient when there are *many* elements to remove, because it only writes the elements we want to keep.
* **Left/Right Pointers:** Highly efficient when there are *few* elements to remove, because it avoids needlessly rewriting valid elements to their current positions.

Since both algorithms operate in O(n) time and O(1) space, deciding which one is "better" in a real-world software environment depends entirely on the distribution of the data you expect to process. For this specific LeetCode problem, the judge accepts both as optimal solutions.
