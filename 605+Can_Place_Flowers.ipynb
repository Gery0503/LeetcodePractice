# 605. Can Place Flowers

[Link to Problem](https://leetcode.com/problems/can-place-flowers/)

### Description

You have a long flowerbed in which some of the plots are planted, and some are not. However, flowers cannot be planted in **adjacent** plots.

Given an integer array `flowerbed` containing `0`s and `1`s, where `0` means empty and `1` means not empty, and an integer `n`, return `true` if `n` new flowers can be planted in the `flowerbed` without violating the no-adjacent-flowers rule and `false` otherwise.

**Example 1:**
```
Input: flowerbed = [1,0,0,0,1], n = 1
Output: true
```

**Example 2:**
```
Input: flowerbed = [1,0,0,0,1], n = 2
Output: false
```

**Constraints:**
- `1 <= flowerbed.length <= 2 * 10^4`
- `flowerbed[i]` is `0` or `1`.
- There are no two adjacent flowers in `flowerbed`.
- `0 <= n <= flowerbed.length`

## My Intuition

_Write down your thoughts, ideas, and insights here._

- **Observations:**
  1. Loop over array, if element is 0, then do verification. 
  2. Verification: if (idx-1) is 1 or (idx+1) is 1 then do nothing and move on
  3. else deduct 1 from n and assign 1 to current plot
- **Edge cases:**
  - one element
  - plant 0 flower
- **Expected approach and complexity:**
  1. Time:  O(n)
  2. Space: O(1)

````python
# history one
from typing import List

def canPlaceFlowers(flowerbed: List[int], n: int) -> bool:
    for i in range(len(flowerbed)):
        if n == 0:
            return True
        if flowerbed[i] != 0 :
            continue

        if (i == 0 or flowerbed[i-1] == 0) and (i == len(flowerbed)-1 or flowerbed[i+1] == 0):
            flowerbed[i]= 1
            n -= 1
        
    return n == 0
            
    
if __name__ == '__main__':
    # Test Case 1
    flowerbed1 = [1, 0, 0, 0, 1]
    n1 = 1
    result1 = canPlaceFlowers(flowerbed1, n1)
    print(f"Test 1 Input: flowerbed = {flowerbed1}, n = {n1}")
    print(f"Test 1 Output: {result1}") # Expected: True

    # Test Case 2
    flowerbed2 = [1, 0, 0, 0, 1]
    n2 = 2
    result2 = canPlaceFlowers(flowerbed2, n2)
    print(f"\nTest 2 Input: flowerbed = {flowerbed2}, n = {n2}")
    print(f"Test 2 Output: {result2}") # Expected: False

    # Test Case 3 (Edge case: planting 0 flowers)
    flowerbed3 = [1, 0, 0, 0, 1]
    n3 = 0
    result3 = canPlaceFlowers(flowerbed3, n3)
    print(f"\nTest 3 Input: flowerbed = {flowerbed3}, n = {n3}")
    print(f"Test 3 Output: {result3}") # Expected: True

    # Test Case 4
    flowerbed4 = [1, 0, 0, 0, 0, 1]
    n4 = 2
    result4 = canPlaceFlowers(flowerbed4, n4)
    print(f"\nTest 4 Input: flowerbed = {flowerbed4}, n = {n4}")
    print(f"Test 4 Output: {result4}") # Expected: False

    # Test Case 5
    flowerbed5 = [0,0,1,0,1]
    n5 = 1
    result5 = canPlaceFlowers(flowerbed5, n5)
    print(f"\nTest 5 Input: flowerbed = {flowerbed5}, n = {n5}")
    print(f"Test 5 Output: {result5}") # Expected: True
````
