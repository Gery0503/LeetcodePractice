# 739. Daily Temperatures

[Link to Problem](https://leetcode.com/problems/daily-temperatures/)

### Description

Given an array of integers `temperatures` represents the daily temperatures, return an array `answer` such that `answer[i]` is the number of days you have to wait after the `i`th day to get a warmer temperature. If there is no future day for which this is possible, keep `answer[i] == 0` instead.

**Example 1:**
```
Input: temperatures = [73,74,75,71,69,72,76,73]
Output: [1,1,4,2,1,1,0,0]
```

**Example 2:**
```
Input: temperatures = [30,40,50,60]
Output: [1,1,1,0]
```

**Example 3:**
```
Input: temperatures = [30,60,90]
Output: [1,1,0]
```

**Constraints:**
- `1 <= temperatures.length <= 10^5`
- `30 <= temperatures[i] <= 100`

## My Intuition

_Write down your thoughts, ideas, and insights here._

- **Observations:**
  1. Use stack to track waiting temperatures index
  2. Use index to calculate waiting days

- **Expected approach and complexity:**
  1. Time:  O(N)
  2. Space: O(N)

````python
from typing import List

# monotonic stack appropach
def dailyTemperatures(temperatures: List[int]) -> List[int]:
    # TODO: Implement your solution here
    stack = []
    result = len(temperatures)*[0]
    for idx, temp in enumerate(temperatures):
        while stack:
            top_idx = stack[-1]
            if temp > temperatures[top_idx]:
                result[top_idx] = idx - top_idx
                stack.pop()
            else:
                break

        stack.append(idx)

    return result

if __name__ == '__main__':
    # Test Case 0
    temperatures0 = [75, 71, 69, 70, 68, 74, 76]
    result0 = dailyTemperatures(temperatures0)
    print(f"Test 0 Input: temperatures = {temperatures0}")
    print(f"Test 0 Output: {result0}") # Expected: [6, 4, 1, 2, 1, 1, 0]

    print()
    # Test Case 1
    temperatures1 = [73, 74, 75, 71, 69, 72, 76, 73]
    result1 = dailyTemperatures(temperatures1)
    print(f"\nTest 1 Input: temperatures = {temperatures1}")
    print(f"Test 1 Output: {result1}") # Expected: [1, 1, 4, 2, 1, 1, 0, 0]

    # Test Case 2
    temperatures2 = [30, 40, 50, 60]
    result2 = dailyTemperatures(temperatures2)
    print(f"\nTest 2 Input: temperatures = {temperatures2}")
    print(f"Test 2 Output: {result2}") # Expected: [1, 1, 1, 0]

    # Test Case 3
    temperatures3 = [30, 60, 90]
    result3 = dailyTemperatures(temperatures3)
    print(f"\nTest 3 Input: temperatures = {temperatures3}")
    print(f"Test 3 Output: {result3}") # Expected: [1, 1, 0]
````

````python
# Enhancement
def dailyTemperatures(temperatures: List[int]) -> List[int]:
    stack = []
    result = [0] * len(temperatures)
    
    for idx, temp in enumerate(temperatures):
        # Resolve all previous days that are colder than the current day
        while stack and temp > temperatures[stack[-1]]:
            prev_idx = stack.pop()
            result[prev_idx] = idx - prev_idx
            
        stack.append(idx)
        
    return result
````
