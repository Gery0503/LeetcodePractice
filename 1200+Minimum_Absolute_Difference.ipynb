# 1200. Minimum Absolute Difference

[Link to Problem](https://leetcode.com/problems/minimum-absolute-difference/) 

### Description

Given an array of distinct integers `arr`, find all pairs of elements with the minimum absolute difference of any two elements. 

A pair `[a, b]` is valid if:
- `a` and `b` are both elements of `arr`,
- `a < b`,
- The absolute difference `|a - b|` is equal to the minimum absolute difference among all possible pairs.

Return a list of all these pairs in ascending order (first by the first element, then by the second element) after sorting `arr` in ascending order. 

**Example 1:**
```
Input: arr = [4, 2, 1, 3]
Output: [[1,2], [2,3], [3,4]]
```

**Example 2:**
```
Input: arr = [1, 3, 6, 10, 15]
Output: [[1,3]]
```


## My Intuition

- Observations:
   1. sort arr
   2. record minimum value and pairs

      
- Expected time/space complexity:
    1. Time : O(nlogn)
    2. Space: O(n)

````python
def minimumAbsDifference(arr):
    arr.sort()
    previous_num = arr[0]
    min_diff_val = None
    pairs = []
    for num in arr[1:]:
        val = num - previous_num
        if not min_diff_val or val < min_diff_val:
            min_diff_val = val
            pairs = [[previous_num, num]]
        elif val == min_diff_val:
            pairs.append([previous_num, num])
        previous_num = num

    return pairs
    
if __name__ == '__main__':
    arr = [4, 2, 1, 3]
    print(minimumAbsDifference(arr))  # replace with your tests after implementation
````

```python
def minimumAbsDifference(arr):
    """
    Return all pairs of integers in arr with the minimum absolute difference.
    arr: list[int] - input array of distinct integers
    Returns: list[list[int]]
    """
    arr.sort()
    min_diff_val = float('inf')
    pairs = []

    for i in range(1, len(arr)):
        diff = arr[i] - arr[i - 1]
        if diff < min_diff_val:
            min_diff_val = diff
            pairs = [[arr[i - 1], arr[i]]]
        elif diff == min_diff_val:
            pairs.append([arr[i - 1], arr[i]])

    return pairs
```
