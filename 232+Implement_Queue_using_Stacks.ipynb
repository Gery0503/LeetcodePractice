### Description
Implement a first in first out (FIFO) queue using only two stacks. The implemented queue should support all the functions of a normal queue (`push`, `peek`, `pop`, and `empty`).

Implement the `MyQueue` class:
- `void push(int x)` Pushes element x to the back of the queue.
- `int pop()` Removes the element from the front of the queue and returns it.
- `int peek()` Returns the element at the front of the queue.
- `boolean empty()` Returns true if the queue is empty, false otherwise.

**Notes:**
- You must use **only** standard operations of a stack, which means only `push to top`, `peek/pop from top`, `size`, and `is empty` operations are valid.
- Depending on your language, your stack implementation may not be limited to using only a stack. You may use a list or deque if it is used as a stack.

---
**Example 1:**
```
Input:
MyQueue myQueue = new MyQueue();
myQueue.push(1);
myQueue.push(2);
myQueue.peek(); // return 1
myQueue.pop();  // return 1
myQueue.empty(); // return false
```

---
**Constraints:**
- `1 <= x <= 9`
- At most `100` calls will be made to `push`, `pop`, `peek`, and `empty`.
- All the calls to `pop` and `peek` are valid.

**Follow-up:** Can you implement the queue such that each operation is **amortized O(1)** time complexity? In other words, performing `n` operations will take overall `O(n)` time even if one of those operations may take longer.

````python
class MyQueue:

    def __init__(self):
        self.s = []
        self.t = []

    def push(self, x: int) -> None:
        if self.t:   # if self.t is not empty
            for _ in range(len(self.t)):   # loop self.t.size()
                self.s.append(self.t.pop())        
        self.s.append(x)

    def pop(self) -> int:
        if self.s:
            for _ in range(len(self.s)):
                self.t.append(self.s.pop())
        return self.t.pop()

    def peek(self) -> int:
        if self.s:
            for _ in range(len(self.s)):
                self.t.append(self.s.pop())
        return self.t[-1]   # get t.top()

    def empty(self) -> bool:
        return not self.s and not self.t
````

### üéØ Step-by-Step Mindset & Approach

This problem is a classic demonstration of using **two stacks to simulate queue behavior** (FIFO using LIFO structures).

### üß† Design Mindset

To simulate a queue:

* `stack1` (`s`) is used to **store input elements** (push side)
* `stack2` (`t`) is used to **reverse the order** (pop/peek side)

#### ‚úÖ Queue Operations Mapping

| Queue Operation | Stack Behavior                               | Implementation             |
| --------------- | -------------------------------------------- | -------------------------- |
| `push(x)`       | Push to `s`                                  | Transfer `t ‚Üí s` then push |
| `pop()`         | Ensure `t` has elements, then `pop` from `t` |                            |
| `peek()`        | Same as pop, but return `t[-1]`              |                            |
| `empty()`       | True if both stacks are empty                |                            |

---

### üîç Why This Approach?

I first implemented a correct but **non-optimal version**: every time I `push`, I transfer all elements from `t` back to `s`. And for `pop` or `peek`, I move everything from `s` back to `t`.

While this is functionally correct, it's **not optimal** for amortized complexity. Every `push`/`pop` becomes O(n) in the worst case.

---

### üõ† Better Approach (Which I Would Propose in a Real Interview):

Instead of flipping stacks each time, we only **transfer when needed**:

````python
class MyQueue:
    def __init__(self):
        self.s_in = []
        self.s_out = []

    def push(self, x: int) -> None:
        self.s_in.append(x)

    def pop(self) -> int:
        self._shift()
        return self.s_out.pop()

    def peek(self) -> int:
        self._shift()
        return self.s_out[-1]

    def empty(self) -> bool:
        return not self.s_in and not self.s_out

    def _shift(self):
        if not self.s_out:
            while self.s_in:
                self.s_out.append(self.s_in.pop())
````

This approach **delays the transfer until absolutely necessary**, making each operation **amortized O(1)**.

---

### ‚è± Complexity Analysis

| Operation | Worst-case | Amortized |
| --------- | ---------- | --------- |
| `push`    | O(1)       | O(1)      |
| `pop`     | O(n)       | O(1)      |
| `peek`    | O(n)       | O(1)      |
| `empty`   | O(1)       | O(1)      |

* Overall: **O(1) amortized time per operation** and **O(n) space**.
