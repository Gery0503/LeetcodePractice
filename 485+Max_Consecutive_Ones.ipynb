# 485. Max Consecutive Ones

[Link to Problem](https://leetcode.com/problems/max-consecutive-ones/)

### Description

Given a binary array `nums`, return _the maximum number of consecutive_ `1`_'s in the array_.

**Example 1:**
```
Input: nums = [1,1,0,1,1,1]
Output: 3
Explanation: The first two digits or the last three digits are consecutive 1s. The maximum number of consecutive 1s is 3.
```

**Example 2:**
```
Input: nums = [1,0,1,1,0,1]
Output: 2
```

**Constraints:**
- `1 <= nums.length <= 10^5`
- `nums[i]` is either `0` or `1`.

## My Intuition

_Write down your thoughts, ideas, and insights here._

- **Observations:**
  1. We could use Greedy approach. Create max_count and current_count
  2. Compare max_num with current_num when it comes to zero
- **Expected approach and complexity:**
  1. Time:  O(n)
  2. Space: O(1)

````python
from typing import List

def findMaxConsecutiveOnes(nums: List[int]) -> int:
    max_count = 0
    current_count = 0
    for num in nums:
        if num == 1:
            current_count += 1
        else:
            max_count = max(max_count, current_count)
            current_count = 0

    return max(max_count, current_count)

if __name__ == '__main__':
    # Test Case 1
    nums1 = [1, 1, 0, 1, 1, 1]
    result1 = findMaxConsecutiveOnes(nums1)
    print(f"Test 1 Input: nums = {nums1}")
    print(f"Test 1 Output: {result1}") # Expected: 3

    # Test Case 2
    nums2 = [1, 0, 1, 1, 0, 1]
    result2 = findMaxConsecutiveOnes(nums2)
    print(f"\nTest 2 Input: nums = {nums2}")
    print(f"Test 2 Output: {result2}") # Expected: 2

    # Test Case 3 (All ones)
    nums3 = [1, 1, 1, 1]
    result3 = findMaxConsecutiveOnes(nums3)
    print(f"\nTest 3 Input: nums = {nums3}")
    print(f"Test 3 Output: {result3}") # Expected: 4
````
