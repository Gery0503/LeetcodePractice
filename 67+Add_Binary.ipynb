# 67. Add Binary

[Link to Problem](https://leetcode.com/problems/add-binary/description/)

### Description
Given two binary strings `a` and `b`, return *their sum as a binary string*.

---
**Example 1:**

Input: `a = "11"`, `b = "1"`
Output: `"100"`

**Example 2:**

Input: `a = "1010"`, `b = "1011"`
Output: `"10101"`

---
**Constraints:**
- `1 <= a.length, b.length <= 10^4`
- `a` and `b` consist only of `'0'` or `'1'` characters.
- Each string does not contain leading zeros except for the zero itself.

```python
class Solution:
    def addBinary(self, a: str, b: str) -> str:
        # Fill 0 to make same length
        if len(a) > len(b):
            b = (len(a)-len(b))*'0' + b
        else:
            a = (len(b)-len(a))*'0' + a

        # Loop start from least significant bit
        bit_result = ''
        carry = 0
        i = len(a)-1
        while i >= 0 :
            bit1 = int(a[i])
            bit2 = int(b[i])
            num = carry + bit1 + bit2
            bit = num % 2
            bit_result = str(bit) + bit_result

            # Determine carry-out bit
            if num >= 2:
                carry = 1
            else:
                carry = 0
                
            i -= 1
        if carry == 1:
            return '1' + bit_result
        return bit_result
# Time: O(n)
# Space: O(n)
```

### 🔍 Problem Summary

We're asked to add two binary numbers represented as strings (`a` and `b`), and return the result as a binary string.

This is conceptually the same as performing **binary addition**, bit by bit, handling carry-overs, similar to how we do decimal addition by hand.

---

### 🧭 Step-by-Step Thought Process

1. **Normalize the Inputs:**

   * We make both strings equal in length by padding the shorter one with leading zeros. This ensures aligned addition from the least significant bit (LSB) to the most significant bit (MSB).
   * This simplifies logic and avoids bounds errors.

2. **Simulate Manual Binary Addition:**

   * Starting from the rightmost bit (LSB), I loop backward.
   * For each bit position:

     * Convert character to integer.
     * Add both bits and the carry from the previous step.
     * The result for the current position is `sum % 2`.
     * Update the carry with `sum // 2`.

3. **Final Carry Handling:**

   * After the loop, if there's a remaining carry (i.e., 1), we prepend it to the result string.

---

### 💡 Why This Approach?

* This mimics how addition is taught at a low level—great for binary arithmetic, and avoids over-reliance on language-specific features like `int(a, 2)` + `int(b, 2)`.
* It ensures better control over overflow, and aligns closely with how systems might implement this at the register level.

---

### ⏱️ Time and Space Complexity

* **Time Complexity:** `O(n)`
  We iterate through the maximum of the lengths of `a` and `b` exactly once.

* **Space Complexity:** `O(n)`
  We store the result as a string with at most `max(len(a), len(b)) + 1` characters.

```python
# Other solutions
class Solution:
  def addBinary(self, a: str, b: str) -> str:
    s = []
    carry = 0
    i = len(a) - 1
    j = len(b) - 1

    while i >= 0 or j >= 0 or carry:
      if i >= 0:
        carry += int(a[i])
        i -= 1
      if j >= 0:
        carry += int(b[j])
        j -= 1
      s.append(str(carry % 2))
      carry //= 2

    return ''.join(reversed(s))
# Time: O(n)
# Space: O(n)
```
