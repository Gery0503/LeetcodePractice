# 258. Add Digits

[Link to Problem](https://leetcode.com/problems/add-digits/)

### Description

Given an integer `num`, repeatedly add all its digits until the result has only one digit, and return it.

**Example 1:**
```
Input: num = 38
Output: 2
Explanation: The process is
38 --> 3 + 8 --> 11
11 --> 1 + 1 --> 2 
Since 2 has only one digit, return it.
```

**Example 2:**
```
Input: num = 0
Output: 0
```

**Constraints:**
- `0 <= num <= 2^31 - 1`

**Follow up:** Could you do it without any loop/recursion in `O(1)` runtime?

## Mathematical Approach

- **Mathematical Approach (Digital Root):** There is a repeating pattern every 9 numbers. The digital root of a number is essentially `num % 9`. 
     - If `num == 0`, the answer is `0`.
     - If `num` is a multiple of 9 (and greater than 0), the answer is `9`.
     - Otherwise, the answer is exactly `num % 9`. 
     - This can be condensed into a single $O(1)$ formula: `1 + (num - 1) % 9` for any `num > 0`.

- **Expected approach and complexity:**
  1. Time: **Math Formula:** $O(1)$. It requires a single modulo operation.
  2. Space: $O(1)$. No extra memory scales with the input size.

````python
# digital root
def addDigits(num: int) -> int:
    if num == 0:
        return 0
    return 1 + (num-1) % 9
    
if __name__ == '__main__':
    # Test Case 1
    num1 = 38
    result1 = addDigits(num1)
    print(f"Test 1 Input: num = {num1}")
    print(f"Test 1 Output: {result1}") # Expected: 2

    # Test Case 2
    num2 = 0
    result2 = addDigits(num2)
    print(f"\nTest 2 Input: num = {num2}")
    print(f"Test 2 Output: {result2}") # Expected: 0
````

## My Intuition

_Write down your thoughts, ideas, and insights here._

- **Observations:**
  1. We can simulate the process by repeatedly using modulo (`% 10`) and division (`// 10`) to extract and sum digits until the number is less than 10.
- **Expected approach and complexity:**
  1. Time: O(log N). The number of iterations depends on the number of digits in $N$, which is proportional to $\log_{10}(N)$. (Note: Because $N$ is capped at $2^{31} - 1$, the loops run a maximum of 3 times, so it strictly bounds to $O(1)$ in practice)
  2. Space: O(1)

````python
# history version: Not the optimal solution
def addDigits(num: int) -> int:
    result = num
    while result >= 10:
        num = result
        result = 0
        while num > 0:
            result += num % 10
            num //= 10
            
    return result
if __name__ == '__main__':
    # Test Case 1
    num1 = 38
    result1 = addDigits(num1)
    print(f"Test 1 Input: num = {num1}")
    print(f"Test 1 Output: {result1}") # Expected: 2

    # Test Case 2
    num2 = 0
    result2 = addDigits(num2)
    print(f"\nTest 2 Input: num = {num2}")
    print(f"Test 2 Output: {result2}") # Expected: 0
````
