# 102. Binary Tree Level Order Traversal

[Link to Problem](https://leetcode.com/problems/binary-tree-level-order-traversal/description/)

### Description
Given the `root` of a binary tree, return the level order traversal of its nodes' values. (i.e., from left to right, level by level).

---
**Example 1:**

Input: `root = [3,9,20,null,null,15,7]`
Output: `[[3],[9,20],[15,7]]`

**Example 2:**

Input: `root = [1]`
Output: `[[1]]`

**Example 3:**

Input: `root = []`
Output: `[]`

---
**Constraints:**
- The number of nodes in the tree is in the range `[0, 2000]`.
- `-1000 <= Node.val <= 1000`

My intuition: Obviously BFS

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

from typing import List
from collections import deque
class Solution:
    def levelOrder(self, root: TreeNode) -> List[List[int]]:
        if not root: return []
        queue = deque([root])
        result = []
        while queue:
            q_size = len(queue)
            values = []
            for _ in range(q_size):
                node = queue.popleft()
                values.append(node.val)
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            result.append(values)
        return result
# Time: O(n)
# Space: O(w)
```

#### üí¨ Implementation Rationale

I chose **Breadth-First Search (BFS)** for this problem because it naturally aligns with the requirement: "return level order traversal." BFS visits nodes level by level, making it a clean and intuitive fit for grouping values at each depth into sublists.

To implement this:

* I initialized a queue using `collections.deque`, which gives O(1) pops from the front.
* I iteratively processed nodes level by level, using a `for` loop to extract exactly the number of nodes at the current level.
* For each node, I stored its value and pushed its children into the queue for the next round.

This method avoids recursion (which could hit stack depth limits on large trees) and maintains a clear separation of levels, which matches the output requirements precisely.

---

#### üßÆ Time and Space Complexity

* **Time Complexity: O(n)**
  Every node is visited exactly once. We spend constant time per node to process it and push children.

* **Space Complexity: O(w)**
  The queue holds at most `w` nodes at a time, where `w` is the maximum width of the tree ‚Äî the number of nodes at the widest level.

```python
class Solution:
    def levelOrder(self, root: TreeNode) -> List[List[int]]:
        ans, level = [], [root]
        while root and level:
            ans.append([node.val for node in level])            
            level = [kid for n in level for kid in (n.left, n.right) if kid]
        return ans
# Time: O(n)
# Space: O(w)
```
