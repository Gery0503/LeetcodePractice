# 804. Unique Morse Code Words

### Description

International Morse Code defines a standard encoding where each letter is mapped to a series of dots and dashes, as follows: `'a'` maps to ".-", `'b'` maps to "-...", `'c'` maps to "-.-.", and so on.

For convenience, the full table for the 26 letters of the English alphabet is given below:
```
[".-","-...","-.-.","-..",".","..-.","--.","....","..",".---","-.-",".-..","--","-.","---",".--.","--.-",".-.","...","-","..-","...-",".--","-..-","-.--","--.."]
```

Now, given an array of strings `words`, each word can be written as a concatenation of the Morse code of each letter. For example, "cab" can be written as "-.-..--...", which is the concatenation of "-.-.", ".-", "-...". Weâ€™ll call such a concatenation the *transformation* of a word.

Return the number of different transformations among all words we have.

**Example 1:**
```
Input: words = ["gin","zen","gig","msg"]
Output: 2
```
### Constraints:

- 1 <= words.length <= 100  
- 1 <= words[i].length <= 12  
- words[i] consists of lowercase English letters.


## My Intuition

_Write your intuition, thoughts, edge cases, and complexity analysis here._

- Observations: use hashmap and Count
- Expected time/space complexity:
  1. Time:  O(n)
  2. Space: O(n)


````python
def uniqueMorseRepresentations(words):
    codes = [".-","-...","-.-.","-..",".","..-.","--.","....","..",".---","-.-",
     ".-..","--","-.","---",".--.","--.-",".-.","...","-","..-","...-",
     ".--","-..-","-.--","--.."]
    alp_to_code = {chr(97+i): code for i, code in enumerate(codes)}
    unique = set()
    count = 0
    for word in words:
        morse = ''
        for letter in word:
            morse += alp_to_code[letter]
        if morse not in unique:
            unique.add(morse)
            count += 1
    return count
            
if __name__ == '__main__':
    words = ["gin", "zen", "gig", "msg"]
    print(uniqueMorseRepresentations(words))  # replace with your own tests after implementation
````
