# 155. Min Stack

[Link to Problem](https://leetcode.com/problems/min-stack/description/)

### Description
Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

Implement the `MinStack` class:
- `MinStack()` initializes the stack object.
- `void push(int val)` pushes the element val onto the stack.
- `void pop()` removes the element on the top of the stack.
- `int top()` gets the top element of the stack.
- `int getMin()` retrieves the minimum element in the stack.

You must implement a solution with `O(1)` time complexity for each function.

---
**Example 1:**

Input:
```
MinStack minStack = new MinStack();
minStack.push(-2);
minStack.push(0);
minStack.push(-3);
minStack.getMin(); // return -3
minStack.pop();
minStack.top();    // return 0
minStack.getMin(); // return -2
```

---
**Constraints:**
- `-2^31 <= val <= 2^31 - 1`
- Methods `pop()`, `top()`, and `getMin()` operations will always be called on **non-empty** stacks.
- At most `3 * 10^4` calls will be made to `push`, `pop`, `top`, and `getMin`.

````python
class MinStack:

    def __init__(self):
        self.stack = []
        self.top_idx = -1

    def push(self, val: int) -> None:
        self.stack += [val]
        self.top_idx += 1

    def pop(self) -> None:
        self.stack = self.stack[:-1]
        self.top_idx -= 1

    def top(self) -> int:
        return self.stack[self.top_idx]

    def getMin(self) -> int:
        return min(self.stack)
````

### ‚úÖ Correct Approach

While the given implementation works functionally, it **does not meet the O(1) requirement** for `getMin()`:

```python
def getMin(self) -> int:
    return min(self.stack)  # ‚ùå O(n)
```

### üí° Optimal Solution (Which I would propose):

Use a **secondary stack (`min_stack`)** to track the minimum value:

````python
class MinStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []  # Stores current min at each level

    def push(self, val: int) -> None:
        self.stack.append(val)
        # Push the minimum between current and new
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)
        else:
            self.min_stack.append(self.min_stack[-1])

    def pop(self) -> None:
        self.stack.pop()
        self.min_stack.pop()

    def top(self) -> int:
        return self.stack[-1]

    def getMin(self) -> int:
        return self.min_stack[-1]
````

| `top`       | O(1)               | O(1)                                  |
| `getMin`    | O(1)               | O(1)                                  |
