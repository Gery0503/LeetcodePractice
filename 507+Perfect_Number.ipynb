# 507. Perfect Number

### Description

A **perfect number** is a positive integer that is equal to the sum of all its **positive divisors**, excluding itself.

A divisor of an integer `x` is a number that divides `x` evenly.

Given an integer `num`, return `true` if `num` is a perfect number, otherwise return `false`.

**Example 1:**
```
Input: num = 28
Output: true
Explanation: 28 = 1 + 2 + 4 + 7 + 14
```

**Example 2:**
```
Input: num = 7
Output: false
```

**Constraints:**
- `1 <= num <= 10^8`

````python
from math import sqrt
def checkPerfectNumber(num):
    if num == 1:
        return False

    div_sum = 1
    limit = int(sqrt(num))

    for i in range(2, limit + 1):
        if num % i == 0:
            div_sum += i
            other = num // i
            if other != i:   # avoid double-counting square root
                div_sum += other

    return div_sum == num
````

```python
def checkPerfectNumber(num):
    # Euclidâ€“Euler theorem:
    # Perfect number = 2^(p-1) * (2^p - 1), where (2^p - 1) is a Mersenne prime

    # All Mersenne primes that generate perfect numbers <= 1e8
    mersenne_p = [2, 3, 5, 7, 13]

    perfect_numbers = {
        (1 << (p - 1)) * ((1 << p) - 1)    # using bit shift: 2^p = (1 << p)
        for p in mersenne_p
    }

    return num in perfect_numbers
```
