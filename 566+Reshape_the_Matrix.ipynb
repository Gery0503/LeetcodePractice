### Description

In MATLAB, there is a handy function called `reshape` which can reshape an m Ã— n matrix into a new one with a different size, while keeping its original data.  
You are given an m Ã— n matrix `mat` and two integers `r` and `c` representing the number of rows and the number of columns of the desired reshaped matrix.  
The reshaped matrix should be filled with all the elements of the original matrix in the same row-traversing order (left to right, top to bottom).  
If the reshape operation with the given parameters is **possible and legal**, output the new reshaped matrix; **otherwise**, output the original matrix. 

**Example 1:**
```
Input: mat = [[1,2],[3,4]], r = 1, c = 4
Output: [[1,2,3,4]]
```

**Example 2:**
```
Input: mat = [[1,2],[3,4]], r = 2, c = 4
Output: [[1,2],[3,4]]
```

### Constraints:

- m == mat.length
- n == mat[i].length
- 1 <= m, n <= 100
- -1000 <= mat[i][j] <= 1000
- 1 <= r, c <= 300


## My Intuition

_Write your intuition, thoughts, edge-cases, and complexity analysis here._

- Observations:
  1. Check if r * c = size(mat)
  2. implement row-traversing order by using two-pointer method
  3. write new mat
- Expected time/space complexity:
  - Time:  O(m * n)
  - Space: O(m * n)


````python
def matrixReshape(mat, r, c):
    m = len(mat)
    n = len(mat[0])
    if m * n != r * c:
        return mat

    i, j = 0, 0  # loop over mat
    new_mat = []
    row = []
    while i < m:
        element = mat[i][j]      
        row.append(element)
        j += 1
        if j == n:
            i += 1
            j = 0
        if len(row) == c:
            new_mat.append(row)
            row = []

    return new_mat
    
if __name__ == '__main__':
    mat = [[1,2],[3,4]]
    r = 4
    c = 1
    print(matrixReshape(mat, r, c))  # replace with your own tests after implementation
````

```python
def matrixReshape(mat, r, c):
    m, n = len(mat), len(mat[0])
    if m * n != r * c:
        return mat

    flat = [num for row in mat for num in row]
    return [flat[i * c:(i + 1) * c] for i in range(r)]


if __name__ == '__main__':
    mat = [[1, 2], [3, 4]]
    print(matrixReshape(mat, 1, 4))  # [[1, 2, 3, 4]]

```
