### Description

The **Fibonacci numbers**, commonly denoted `F(n)` form a sequence, called the **Fibonacci sequence**, such that each number is the sum of the two preceding ones, starting from `0` and `1`. That is,

```
F(0) = 0, F(1) = 1
F(n) = F(n - 1) + F(n - 2), for n > 1.
```

Given `n`, calculate `F(n)`.

**Example 1:**
```
Input: n = 2
Output: 1
Explanation: F(2) = F(1) + F(0) = 1 + 0 = 1.
```

**Example 2:**
```
Input: n = 3
Output: 2
Explanation: F(3) = F(2) + F(1) = 1 + 1 = 2.
```

**Example 3:**
```
Input: n = 4
Output: 3
Explanation: F(4) = F(3) + F(2) = 2 + 1 = 3.
```

**Constraints:**
- `0 <= n <= 30`

## My Intuition

_Write down your thoughts, ideas, and insights here._

- **Observations:**
  1. Use dynamic programming
  2. Recursive approach is easy but takes more time/space complexity
- **Edge cases:**
  - n = 0 or n = 1
- **Expected approach and complexity:**
  1. Time:  O(n)
  2. Space: O(1)

````python
def fib(n: int) -> int:
    a = 0
    b = 1
    for _ in range(n):
        a, b = b, a+b
        
    return a

if __name__ == '__main__':
    # Test Case 1
    n1 = 2
    result1 = fib(n1)
    print(f"Test 1 Input: n = {n1}")
    print(f"Test 1 Output: {result1}") # Expected: 1

    # Test Case 2
    n2 = 3
    result2 = fib(n2)
    print(f"\nTest 2 Input: n = {n2}")
    print(f"Test 2 Output: {result2}") # Expected: 2

    # Test Case 3
    n3 = 4
    result3 = fib(n3)
    print(f"\nTest 3 Input: n = {n3}")
    print(f"Test 3 Output: {result3}") # Expected: 3
````
