### Description

Given an integer `rowIndex`, return the `rowIndex`th (0-indexed) row of the Pascal's triangle.

In Pascal's triangle, each number is the sum of the two numbers directly above it, as shown:

```
      [1]    (Row 0)
    [1] [1]  (Row 1)
  [1] [2] [1]  (Row 2)
 [1] [3] [3] [1]  (Row 3)
[1] [4] [6] [4] [1]  (Row 4)
```

**Example 1:**
```
Input: rowIndex = 3
Output: [1,3,3,1]
```

**Example 2:**
```
Input: rowIndex = 0
Output: [1]
```

**Example 3:**
```
Input: rowIndex = 1
Output: [1,1]
```

**Constraints:**
- `0 <= rowIndex <= 33`

**Follow up:** Could you optimize your algorithm to use only $O(k)$ extra space, where $k$ is the given `rowIndex`?

## My Intuition

- **Observations:**
  1. The problem asks for a single row, unlike the previous problem which asked for all rows up to `numRows`.
  2. We need to calculate the row using the elements from the previous (or current) row.
- **Edge cases:**
  - What if rowIndex is 0?
- **Expected approach and complexity (with $O(k)$ space challenge):**
  1. Time: O(k**2)
  2. Space: O(k)

````python
# Recusive approach (Space: O(k**2))
from typing import List

def getRow(rowIndex: int) -> List[int]:
    if rowIndex == 0:
        return [1]
    
    previous_row = getRow(rowIndex-1)
    current_row = [1] * (rowIndex + 1)
    for i in range(1, rowIndex):
        current_row[i] = previous_row[i - 1] + previous_row[i]
    
    return current_row
    

if __name__ == '__main__':
    # Test Case 1
    rowIndex1 = 3
    result1 = getRow(rowIndex1)
    print(f"Input: {rowIndex1}")
    print(f"Output: {result1}") # Expected: [1,3,3,1]

    # Test Case 2
    rowIndex2 = 0
    result2 = getRow(rowIndex2)
    print(f"Input: {rowIndex2}")
    print(f"Output: {result2}") # Expected: [1]
````

  * **Analysis:**
      * **Auxiliary Space (excluding recursion):** $O(\text{rowIndex})$ for storing `current_row` and `previous_row` at each level.
      * **Recursion Stack Space:** $O(\text{rowIndex})$ depth, where each call stores its local variables (`previous_row`, `current_row`).
      * **Total Space:** The `previous_row` list returned by each sub-call is kept in memory by the caller. Since the lengths are $1, 2, \dots, \text{rowIndex}+1$, the total space used is $O(\text{rowIndex}^2)$. This **does not meet the $O(k)$ extra space constraint.**

### ðŸ’¡ Suggested Improvement: Iterative $O(k)$ Space Solution

To meet the $O(k)$ space constraint, we need an **iterative** approach that reuses a single list.

The key idea is to calculate the row *in-place* starting from the second element from the right.

For a row $R$:

  * $R[i] = R_{\text{previous}}[i-1] + R_{\text{previous}}[i]$

If we only use one list, say `current_row`, and update it from left to right, when we calculate $current\_row[i]$, we will overwrite $current\_row[i-1]$ *before* we need its original value for the next calculation.

**Solution:** Iterate from **right to left** (from $i$ up to 1). When calculating the new value for $R[i]$, both $R[i-1]$ and $R[i]$ still hold the correct values from the previous row calculation.

**Revised Iterative Code (Meeting $O(k)$ Space):**

```python
def getRow_optimized(rowIndex: int) -> List[int]:
    # Initialize the row with all 1s (Row 0 is [1], Row 1 is [1, 1], etc.)
    # The length of the rowIndex-th row is rowIndex + 1.
    result = [1] * (rowIndex + 1)
    
    # Start loop from row 2 (r=2) up to the target rowIndex (r=rowIndex)
    # Row 0 and Row 1 are base cases handled by the initialization of result.
    for r in range(2, rowIndex + 1):
        # Calculate elements for row 'r' using the current 'result' list
        # which represents the previous row (r-1).
        # We must iterate from right to left (i from r-1 down to 1) 
        # to use the previous row's values before they are overwritten.
        
        # r is the length of the current row minus 1 (index of the last element, which is 1)
        # We stop at index 1 because result[0] is always 1.
        for i in range(r - 1, 0, -1):
            # Pascal's rule: element = previous[i-1] + previous[i]
            # Since we iterate right-to-left:
            # result[i-1] holds previous[i-1]
            # result[i] holds previous[i]
            result[i] = result[i - 1] + result[i]
            
    return result

```

````python
# Binomial coefficient

from typing import List

class Solution:
    """
    Calculates the rowIndex-th row of Pascal's Triangle using the binomial coefficient formula.
    
    Time Complexity: O(k), where k is rowIndex. This is optimal as we must generate k+1 elements.
    Space Complexity: O(k) for storing the result list, satisfying the follow-up constraint.
    """
    def getRow(self, rowIndex: int) -> List[int]:
        # The first element C(n, 0) is always 1.
        res = [1]
        
        # 'prev' stores the previous binomial coefficient, C(n, k-1).
        # It starts as C(n, 0) = 1.
        prev = 1
        
        # Iterate from k=1 up to k=rowIndex (since C(n, n) is the last element).
        # We start at k=1 because C(n, 0) is already added.
        for k in range(1, rowIndex + 1):
            # Calculate the next binomial coefficient, C(n, k), using the previous one:
            # C(n, k) = C(n, k-1) * (n - k + 1) / k
            # Note: Integer division '//' is mandatory here since the result must be an integer.
            # Python's integers handle the large numbers (up to rowIndex=33) without overflow.
            next_val = prev * (rowIndex - k + 1) // k
            
            res.append(next_val)
            
            # Update 'prev' to be ready for the next iteration.
            prev = next_val
            
        return res

# --- Example Usage ---
if __name__ == '__main__':
    solver = Solution()
    
    # Test Case 1: rowIndex = 3 -> [1, 3, 3, 1]
    row3 = solver.getRow(3)
    print(f"Row 3: {row3}")
    
    # Test Case 2: rowIndex = 0 -> [1]
    row0 = solver.getRow(0)
    print(f"Row 0: {row0}")

    # Test Case 3: rowIndex = 5 -> [1, 5, 10, 10, 5, 1]
    row5 = solver.getRow(5)
    print(f"Row 5: {row5}")
````
