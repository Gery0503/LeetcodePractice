### Description

Reverse bits of a given 32 bits unsigned integer.

Note:

Note that in some languages, such as Java, there is no unsigned integer type. In this case, both input and output will be given as a signed integer type. They should not affect your implementation, as the integer's internal binary representation is the same, whether it is signed or unsigned.
In Java, the compiler represents the signed integers using 2's complement notation. Therefore, in Example 2 below, the input represents the signed integer -3 and the output represents the signed integer -1073741825.

---
**Example 1:**

Input: `n = 43261596`

Output: `964176192`

Explanation:

Integer	|Binary
-------|------
43261596	|00000010100101000001111010011100
964176192	|00111001011110000010100101000000

**Example 2:**


Input: n = `2147483644`

Output: `1073741822`

Explanation:

Integer	|Binary
--------|--------
2147483644	|01111111111111111111111111111100
1073741822	|00111111111111111111111111111110

---
**Constraints:**
- 0 <= n <= $2^{31}$ - 2
- n is even.

**Follow up:**
If this function is called many times, how would you optimize it?

My intuition: use bit manipulation

````python
# My solution
class Solution:
    def reverseBits(self, n: int) -> int:
        result = 0
        for _ in range(32):
            least_bit = n & 1
            result = result*2 + least_bit
            n >>= 1
        return result
# Time: O(1)
# Space: O(1)
````

````python
# Refactor
class Solution:
    def reverseBits(self, n: int) -> int:
        result = 0
        for _ in range(32):
            result = (result << 1) | (n & 1) 
            n >>= 1
        return result
# Time: O(1)
# Space: O(1)
````

````python
# Leetcode solution (Bitmask Swapping)
class Solution:
    def reverseBits(self, n: int) -> int:
        n = ((n & 0xffff0000) >> 16) | ((n & 0x0000ffff) << 16)
        n = ((n & 0xff00ff00) >> 8) | ((n & 0x00ff00ff) << 8)
        n = ((n & 0xf0f0f0f0) >> 4) | ((n & 0x0f0f0f0f) << 4)
        n = ((n & 0xcccccccc) >> 2) | ((n & 0x33333333) << 2)
        n = ((n & 0xaaaaaaaa) >> 1) | ((n & 0x55555555) << 1)
        
        return n
# Time: O(1)
# Space: O(1)
````

### 🔁 The Problem Recap

---

### ⚡ Why Lookup Table?

Your original bit-reversing solution is **O(1)** in time (because it's fixed 32 iterations), but:

> 🔁 When the function is called **many times**, say millions, we want to **minimize repeated computation**.

That’s where **precomputing with a lookup table** becomes powerful.

---

### 💡 Lookup Table Idea

Split the 32-bit integer into **4 chunks of 8 bits**, reverse each chunk using a precomputed table of 256 entries (i.e., all possible 8-bit values), then recombine them.

Why 8-bit chunks?
Because:

* 8-bit values = `0 to 255` → only **256 possible combinations**
* We can reverse each byte using a lookup in **O(1)** time

---

### ✅ Step-by-Step

#### Step 1: Precompute the reverse of 8-bit numbers

```python
LOOKUP = [0] * 256
for i in range(256):
    b = i
    rev = 0
    for _ in range(8):
        rev = (rev << 1) | (b & 1)
        b >>= 1
    LOOKUP[i] = rev
```

#### Step 2: Use the table in reverseBits

```python
class Solution:
    def reverseBits(self, n: int) -> int:
        LOOKUP = self.lookup
        return (
            (LOOKUP[n & 0xff] << 24) |
            (LOOKUP[(n >> 8) & 0xff] << 16) |
            (LOOKUP[(n >> 16) & 0xff] << 8) |
            (LOOKUP[(n >> 24) & 0xff])
        )

    def __init__(self):
        self.lookup = [0] * 256
        for i in range(256):
            b = i
            rev = 0
            for _ in range(8):
                rev = (rev << 1) | (b & 1)
                b >>= 1
            self.lookup[i] = rev
```

---

### 🧠 Why This Is Efficient

* **Time complexity:** O(1) — still constant, but with smaller constant factor due to only 4 operations instead of 32 bit shifts.
* **Space complexity:** O(256) = O(1) — only one lookup table in memory
* **Scalability:** When reversing **millions of integers**, this method drastically reduces CPU cycles per call.

### Why not split the 32-bit integer into 2 chunks of 16 bits?

Excellent and insightful question!

You **can** split a 32-bit integer into **two 16-bit chunks** instead of four 8-bit chunks, but let me walk you through **why we typically choose 8-bit chunks (1 byte)** over 16-bit chunks, and what the trade-offs are.

---

### 🔍 Option 1: **Split into four 8-bit chunks**

* 8-bit has `2^8 = 256` combinations.
* Lookup table size: **256 entries** → very small, easily fits in L1/L2 CPU cache.
* Easy bitwise operations for shifting and combining.
* Memory-efficient and extremely fast.

---

### 🔍 Option 2: **Split into two 16-bit chunks**

* 16-bit has `2^16 = 65,536` combinations.
* Lookup table size: **65,536 entries** → 256x larger than the 8-bit table.
* Takes \~256 KB memory (vs. 256 bytes for 8-bit) → **higher cache miss risk**.
* Lookup is still O(1), but performance **can degrade due to cache pressure**, especially in memory-constrained environments or when this function is called frequently in large-scale systems.

---

### ✅ So Why Prefer 8-bit?

| Criteria           | 8-bit Table      | 16-bit Table                             |
| ------------------ | ---------------- | ---------------------------------------- |
| **Speed**          | Very fast        | Slightly faster per call (less shifting) |
| **Memory Use**     | 256 B            | 256 KB (1000x bigger)                    |
| **Cache Friendly** | ✅ Fits in cache  | ❌ Risky for L1/L2                        |
| **Simplicity**     | Easy composition | Slightly more complex                    |
| **Scalability**    | More scalable    | Not ideal for embedded / low-mem         |

---

### 👨‍🔧 Final Thoughts

* **16-bit chunking is theoretically faster** (fewer shifts and lookups), but **not practically worth it** unless you're in a niche domain where performance per cycle matters and memory is abundant (e.g., GPU kernels or DSPs).
* For most real-world applications — and especially for interviews — the 8-bit table is the **sweet spot between speed and space**.

````python
import time

# 8-bit reverse lookup table
reverse8 = [0] * 256
for i in range(256):
    b = i
    rev = 0
    for _ in range(8):
        rev = (rev << 1) | (b & 1)
        b >>= 1
    reverse8[i] = rev

# 16-bit reverse lookup table
reverse16 = [0] * 65536
for i in range(65536):
    b = i
    rev = 0
    for _ in range(16):
        rev = (rev << 1) | (b & 1)
        b >>= 1
    reverse16[i] = rev

# 8-bit based reversal
def reverseBits8(n: int) -> int:
    return (
        (reverse8[n & 0xff] << 24) |
        (reverse8[(n >> 8) & 0xff] << 16) |
        (reverse8[(n >> 16) & 0xff] << 8) |
        (reverse8[(n >> 24) & 0xff])
    )

# 16-bit based reversal
def reverseBits16(n: int) -> int:
    return (
        (reverse16[n & 0xffff] << 16) |
        (reverse16[(n >> 16) & 0xffff])
    )

# Benchmarking
def benchmark(func, iterations=1000000):
    start = time.time()
    for i in range(iterations):
        func(i)
    end = time.time()
    return end - start

# Run benchmarks
time_8bit = benchmark(reverseBits8)
time_16bit = benchmark(reverseBits16)

time_8bit, time_16bit
````

When comparing the two optimized approaches for reversing bits using lookup tables:

### ⏱ Benchmark Results:

* **8-bit lookup table approach:** \~0.65 seconds
* **16-bit lookup table approach:** \~0.28 seconds
  ✅ **Faster by more than 2x**

---

### 💡 Why is the 16-bit version faster?

1. **Fewer table lookups:**

   * 8-bit version does 4 lookups (1 for each byte).
   * 16-bit version does only 2 lookups (1 for each half-word).

2. **Fewer bitwise operations:**

   * Less shifting and masking are needed.
   * Less combining and rearranging work in the final result.

3. **Cache locality:**

   * Although the 16-bit table (65,536 entries) is larger than the 8-bit one (256 entries), modern CPUs can still handle such lookup tables efficiently in L2/L3 cache.

---

### 🧠 When to use 8-bit vs 16-bit:

| Strategy      | Lookup Table Size | Speed                | Use Case                     |
| ------------- | ----------------- | -------------------- | ---------------------------- |
| 8-bit chunks  | 256 entries       | Slower               | Lower memory footprint       |
| 16-bit chunks | 65,536 entries    | Significantly faster | When performance is critical |

---

### ✅ Summary:

Splitting into **2 chunks of 16 bits is preferred** over 4 chunks of 8 bits in terms of **runtime performance**, because it:

* Reduces the number of lookups,
* Minimizes shift operations,
* Leverages modern CPU cache capabilities.

If you're optimizing for **speed over memory**, go with **16-bit** chunking.
