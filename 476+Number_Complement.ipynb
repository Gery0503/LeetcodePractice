# 476. Number Complement

[Link to Problem](https://leetcode.com/problems/number-complement/)

### Description

The complement of an integer is the integer you get when you flip all the 0’s to 1’s and all the 1’s to 0’s in its binary representation. 

For example, the integer 5 is "101" in binary and its complement is "010", which is the integer 2. 

Given a positive integer `num`, return its complement number. 

**Example 1:**
```
Input: num = 5
Output: 2
```

**Example 2:**
```
Input: num = 1
Output: 0
```

### Constraint
- 1 <= num < 2**31

## My Intuition

_Write your intuition, thoughts, edge cases, and complexity analysis here._

- Observations:
  1. use XOR 1 to get complement
- Expected time/space complexity:
  - Time:  O(len(digits))
  - Space: O(1)


````python
def findComplement(num):
    result = 0
    shift = 0
    while num != 0:
        rightmost = num % 2
        num = num // 2
        complement = rightmost ^ 1
        result += complement*(2**shift)
        shift += 1

    return result
    
if __name__ == '__main__':
    num = 5
    print(findComplement(num))  # replace with your own tests after implementation
````
